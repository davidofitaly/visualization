Notatka 1 Przygotowanie danych.



```python
"""
Notatka: Wczytywanie i wyświetlanie danych zbioru Iris

Poniżej znajduje się kod do wczytania danych zbioru Iris przy użyciu biblioteki Seaborn oraz wyświetlenia trzech pierwszych wierszy ze zbioru danych.

1. Wczytywanie danych zbioru Iris:
   - `df = sns.load_dataset('iris')`: Wczytuje zbiór danych Iris przy użyciu funkcji `load_dataset()` z biblioteki Seaborn i zapisuje go do zmiennej `df`. Zbiór ten zawiera informacje o trzech gatunkach irysów: setosa, versicolor i virginica.

2. Wyświetlanie trzech pierwszych wierszy danych:
   - `df.head(3)`: Wyświetla trzy pierwsze wiersze zbioru danych, co pozwala zapoznać się z jego strukturą oraz zawartością.

Kod ten służy do wczytania danych zbioru Iris i ich wstępnego przeglądu.
"""

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df = sns.load_dataset('iris')
df.head(3)
```

Notatka 2 Stworzenie wykresu pairplot().

```python
"""
Notatka: Tworzenie wykresu typu Pairplot dla zbioru danych Iris

Poniżej znajduje się kod tworzący wykres typu Pairplot dla zbioru danych Iris przy użyciu biblioteki Seaborn. Pairplot pozwala na wizualizację rozkładu danych oraz zależności między parami zmiennych numerycznych.

1. Konfiguracja stylu:
   - `sns.set()`: Ustawia domyślny styl wykresów.

2. Tworzenie wykresu Pairplot:
   - `sns.pairplot(df)`: Tworzy wykres typu Pairplot dla zbioru danych `df`. Wykres ten przedstawia histogramy na przekątnych oraz wykresy punktowe dla wszystkich par zmiennych numerycznych w zbiorze danych.

Pairplot jest przydatnym narzędziem do wstępnej analizy danych, pozwalając na szybkie zrozumienie ich struktury oraz identyfikację potencjalnych zależności między zmiennymi.
"""

sns.set()
sns.pairplot(df)
```

```python
"""
Notatka: Tworzenie wykresu typu Pairplot dla zbioru danych Iris z podziałem na gatunki

Poniżej znajduje się kod tworzący wykres typu Pairplot dla zbioru danych Iris z podziałem na gatunki przy użyciu biblioteki Seaborn. Pairplot pozwala na wizualizację rozkładu danych oraz zależności między parami zmiennych numerycznych.

1. Konfiguracja stylu:
   - `sns.set()`: Ustawia domyślny styl wykresów.

2. Tworzenie wykresu Pairplot z podziałem na gatunki:
   - `sns.pairplot(df, hue='species', palette='pastel')`: Tworzy wykres typu Pairplot dla zbioru danych `df`, gdzie dane są podzielone na gatunki. Wykres ten przedstawia histogramy na przekątnych oraz wykresy punktowe dla wszystkich par zmiennych numerycznych w zbiorze danych, z różnicowaniem kolorów w zależności od gatunku.

Pairplot z podziałem na gatunki umożliwia lepsze zrozumienie zróżnicowania danych między poszczególnymi gatunkami irysów.
"""

sns.set()
sns.pairplot(df, hue='species', palette='pastel')
```

```python
"""
Notatka: Tworzenie wykresu typu PairGrid dla zbioru danych Iris z podziałem na gatunki

Poniżej znajduje się kod tworzący wykres typu PairGrid dla zbioru danych Iris z podziałem na gatunki, gdzie na przekątnych znajdują się wykresy gęstościowe (KDE), a poza przekątnymi wykresy KDE z różnymi poziomami linii konturowych.

1. Tworzenie obiektu PairGrid:
   - `g = sns.PairGrid(df, hue='species')`: Tworzy obiekt PairGrid dla zbioru danych `df`, gdzie dane są podzielone na gatunki.

2. Mapowanie wykresów gęstościowych na przekątne:
   - `g.map_diag(sns.kdeplot)`: Mapuje wykresy gęstościowe (KDE) na przekątne wykresów.

3. Mapowanie wykresów KDE z poziomami linii konturowych poza przekątnymi:
   - `g.map_offdiag(sns.kdeplot, nlevels=6)`: Mapuje wykresy gęstościowe (KDE) z różnymi poziomami linii konturowych poza przekątnymi wykresów.
"""

g = sns.PairGrid(df, hue='species')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, nlevels=6)
```
