```python
"""
Notatka: Import danych dotyczących diamentów za pomocą biblioteki Seaborn

Ten skrypt zawiera kod do zaimportowania danych dotyczących diamentów za pomocą funkcji `load_dataset()` z biblioteki Seaborn. Dane te są następnie przypisane do zmiennej `df` i wyświetlone za pomocą metody `head()`.

1. Importowanie biblioteki Seaborn:
   - `import seaborn as sns`: Importuje bibliotekę Seaborn i nadaje jej alias `sns`.

2. Zaimportowanie danych dotyczących diamentów:
   - `df = sns.load_dataset('diamonds')`: Ładuje zestaw danych dotyczących diamentów za pomocą funkcji `load_dataset()` z biblioteki Seaborn i przypisuje je do zmiennej `df`.

3. Wyświetlenie początkowych wierszy danych:
   - `df.head()`: Wyświetla pierwsze pięć wierszy danych z ramki danych `df`, aby zapoznać się z ich strukturą i zawartością.

Dane te zawierają różne cechy diamentów, takie jak masa, kolor, przepuszczalność światła, a także ich ceny.
"""

import seaborn as sns
df = sns.load_dataset('diamonds')
df.head()
```

```python
df.describe()
```

```python
df = df.cut.value_counts()
df = df.reset_index()
df
```

```python
"""
Notatka: Tworzenie wykresu kołowego dla danych dotyczących rodzaju cięcia diamentów

Ten skrypt zawiera kod do tworzenia wykresu kołowego za pomocą biblioteki Plotly Graph Objects dla danych dotyczących rodzaju cięcia diamentów.

1. Tworzenie wykresu kołowego:
   - `fig = go.Figure(data=[go.Pie(labels=df.index, values=df.cut)])`
     - Tworzy wykres kołowy na podstawie danych z ramki danych `df`, gdzie etykiety (labels) są pobierane z indeksu ramki danych, a wartości (values) są pobierane z kolumny `cut` (rodzaj cięcia diamentu).
     - Parametry:
       - `labels=df.index`: Etykiety (nazwy kategorii) pobierane z indeksu ramki danych `df`.
       - `values=df.cut`: Wartości (procentowy udział) pobierane z kolumny `cut` (rodzaj cięcia diamentu).

2. Wyświetlenie wykresu:
   - `fig.show()`: Wyświetla wykres kołowy.

Wykres kołowy prezentuje procentowy udział różnych rodzajów cięć diamentów w zbiorze danych.
"""

import plotly.graph_objects as go

fig = go.Figure(data=[go.Pie(labels=df.index, values=df.cut)])
fig.show()
```

```python
"""
Notatka: Tworzenie wykresu kołowego z wewnętrznym otworem dla danych dotyczących rodzaju cięcia diamentów

Ten skrypt zawiera kod do tworzenia wykresu kołowego z wewnętrznym otworem za pomocą biblioteki Plotly Graph Objects dla danych dotyczących rodzaju cięcia diamentów.

1. Tworzenie wykresu kołowego z wewnętrznym otworem:
   - `fig = go.Figure(data=[go.Pie(labels=df.index, values=df.cut, hole=0.4)])`
     - Tworzy wykres kołowy z wewnętrznym otworem na podstawie danych z ramki danych `df`, gdzie etykiety (labels) są pobierane z indeksu ramki danych, a wartości (values) są pobierane z kolumny `cut` (rodzaj cięcia diamentu).
     - Parametry:
       - `labels=df.index`: Etykiety (nazwy kategorii) pobierane z indeksu ramki danych `df`.
       - `values=df.cut`: Wartości (procentowy udział) pobierane z kolumny `cut` (rodzaj cięcia diamentu).
       - `hole=0.4`: Określa wielkość wewnętrznego otworu w wykresie kołowym. Wartość 0.4 oznacza, że otwór będzie stanowił 40% średnicy całego wykresu.

2. Wyświetlenie wykresu:
   - `fig.show()`: Wyświetla wykres kołowy z wewnętrznym otworem.

Wykres kołowy z wewnętrznym otworem pozwala na dodanie dodatkowej informacji do prezentacji danych, zachowując jednocześnie czytelność i estetykę.
"""

fig = go.Figure(data=[go.Pie(labels=df.index, values=df.cut, hole=0.4)])
fig.show()
```
