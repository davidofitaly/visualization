```python
"""
Notatka: Załadowanie zestawu danych dotyczących lotów

Ten skrypt zawiera kod do wczytania zestawu danych dotyczących lotów za pomocą funkcji `load_dataset()` z biblioteki Seaborn. Następnie wyświetla pierwsze kilka wierszy tego zestawu danych.

1. Załaduj zestaw danych:
   - `df = load_dataset('flights')`: Wczytuje zestaw danych dotyczących lotów za pomocą funkcji `load_dataset()` z biblioteki Seaborn i przypisuje go do zmiennej `df`.

2. Wyświetlenie pierwszych wierszy zestawu danych:
   - `df.head()`: Wyświetla pierwsze kilka wierszy wczytanego zestawu danych, co pozwala na szybką wstępną analizę danych.
"""

from seaborn import load_dataset
import pandas as pd
import plotly.express as px

df = load_dataset('flights')
df.head()
```

```python
df.describe()
```

```python
"""
Notatka: Tworzenie wykresu słupkowego dla danych dotyczących liczby pasażerów w zależności od roku i miesiąca

Ten skrypt zawiera kod do tworzenia wykresu słupkowego za pomocą funkcji `px.bar()` z biblioteki Plotly Express dla danych dotyczących liczby pasażerów w zależności od roku i miesiąca.

1. Tworzenie wykresu słupkowego:
   - `px.bar(df, x='year', y='passengers', title='Number of passengers by year', width=1000, color='month')`
     - Tworzy wykres słupkowy na podstawie danych z ramki danych `df`, gdzie wartości na osi X są pobierane z kolumny `year` (rok), a wartości na osi Y są pobierane z kolumny `passengers` (liczba pasażerów).
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `x='year'`: Nazwa kolumny reprezentującej wartości na osi X (rok).
       - `y='passengers'`: Nazwa kolumny reprezentującej wartości na osi Y (liczba pasażerów).
       - `title='Number of passengers by year'`: Tytuł wykresu.
       - `width=1000`: Szerokość wykresu w pikselach.
       - `color='month'`: Kolumna, według której kolorowane są słupki (miesiąc).
"""

px.bar(df, x='year',
       y='passengers',
       title='Number of passengers by year',
       width=1000,
       color='month')
```

```python
"""
Notatka: Tworzenie wykresu słupkowego dla danych dotyczących liczby pasażerów w zależności od miesiąca i roku

Ten skrypt zawiera kod do tworzenia wykresu słupkowego za pomocą funkcji `px.bar()` z biblioteki Plotly Express dla danych dotyczących liczby pasażerów w zależności od miesiąca i roku.

1. Tworzenie wykresu słupkowego:
   - `px.bar(df, x='month', y='passengers', title='Number of passengers by month', width=1000, color='year')`
     - Tworzy wykres słupkowy na podstawie danych z ramki danych `df`, gdzie wartości na osi X są pobierane z kolumny `month` (miesiąc), a wartości na osi Y są pobierane z kolumny `passengers` (liczba pasażerów).
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `x='month'`: Nazwa kolumny reprezentującej wartości na osi X (miesiąc).
       - `y='passengers'`: Nazwa kolumny reprezentującej wartości na osi Y (liczba pasażerów).
       - `title='Number of passengers by month'`: Tytuł wykresu.
       - `width=1000`: Szerokość wykresu w pikselach.
       - `color='year'`: Kolumna, według której kolorowane są słupki (rok).
"""

px.bar(df, x='month',
       y='passengers',
       title='Number of passengers by month',
       width=1000,
       color='year')
```

```python
"""
Notatka: Tworzenie wykresu słupkowego poziomego dla danych dotyczących liczby pasażerów w zależności od miesiąca i roku

Ten skrypt zawiera kod do tworzenia wykresu słupkowego poziomego za pomocą funkcji `px.bar()` z biblioteki Plotly Express dla danych dotyczących liczby pasażerów w zależności od miesiąca i roku.

1. Tworzenie wykresu słupkowego poziomego:
   - `px.bar(df, y='month', x='passengers', title='Number of passengers by month', width=1000, color='year', orientation='h')`
     - Tworzy wykres słupkowy poziomy na podstawie danych z ramki danych `df`, gdzie wartości na osi Y są pobierane z kolumny `month` (miesiąc), a wartości na osi X są pobierane z kolumny `passengers` (liczba pasażerów).
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `y='month'`: Nazwa kolumny reprezentującej wartości na osi Y (miesiąc).
       - `x='passengers'`: Nazwa kolumny reprezentującej wartości na osi X (liczba pasażerów).
       - `title='Number of passengers by month'`: Tytuł wykresu.
       - `width=1000`: Szerokość wykresu w pikselach.
       - `color='year'`: Kolumna, według której kolorowane są słupki (rok).
       - `orientation='h'`: Ustawia orientację wykresu na poziomą.
"""

px.bar(df, y='month',
       x='passengers',
       title='Number of passengers by month',
       width=1000,
       color='year',
       orientation='h')
```
