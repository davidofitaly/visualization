Notatka 1.1 Rodzaje wykresów Seaborn.

```python
"""
Notatka 1.1: Rodzaje wykresów w bibliotece Seaborn i ich zastosowania

Biblioteka Seaborn oferuje wiele rodzajów wykresów, które są przydatne do analizy danych i wizualizacji.
Poniżej przedstawiono kilka popularnych rodzajów wykresów dostępnych w Seaborn oraz ich główne zastosowania:

1. Wykres liniowy (relplot lub lineplot):
   - Zastosowanie: Służy do przedstawiania zmiany wartości zmiennej w czasie lub w zależności od innej zmiennej. Idealny do analizy trendów i porównań.

2. Wykres punktowy (relplot lub scatterplot):
   - Zastosowanie: Wykorzystywany do przedstawiania punktów danych na płaszczyźnie, gdzie każdy punkt może być różny pod względem koloru, rozmiaru lub kształtu.
   Doskonały do analizy zależności między dwiema zmiennymi.

3. Wykres słupkowy (catplot lub barplot):
   - Zastosowanie: Nadaje się do porównywania wielkości lub ilości między różnymi kategoriami. Może być wykorzystany do prezentacji danych kategorycznych.

4. Wykres kołowy (Nie ma bezpośredniego odpowiednika w Seaborn):
   - Zastosowanie: Bardziej popularny w bibliotece Matplotlib. Idealny do pokazywania udziałów procentowych poszczególnych kategorii w całkowitej wartości.

5. Histogram (histplot):
   - Zastosowanie: Wykorzystywany do analizy rozkładu danych numerycznych poprzez pokazanie częstotliwości występowania wartości w określonych przedziałach.

6. Wykres pudełkowy (boxplot):
   - Zastosowanie: Służy do przedstawiania rozkładu danych numerycznych wraz z wykrywaniem wartości odstających, mediany, kwartylów itp.
   Przydatny do analizy rozkładu danych i porównań między grupami.

7. Wykres powierzchniowy (Nie ma bezpośredniego odpowiednika w Seaborn):
   - Zastosowanie: W Seabornie nie ma funkcji do rysowania wykresów powierzchniowych.
   W przypadku wizualizacji danych trójwymiarowych na powierzchniach zalecane jest użycie innych narzędzi.

8. Wykres heatmap (heatmap):
   - Zastosowanie: Idealny do przedstawiania danych w formie kolorowej mapy, gdzie intensywność kolorów reprezentuje wartości numeryczne w określonych punktach.

Wybór odpowiedniego rodzaju wykresu zależy od rodzaju danych, które chcesz przedstawić, oraz celu analizy.
Dobrze dobrane wykresy mogą pomóc w efektywnej komunikacji informacji i lepszym zrozumieniu danych przez odbiorcę.
"""
```

Notatka 1.2 Stworzenie wykresu sns.replot().

```python
"""
Notatka 1.2: Wykres rozrzutu na podstawie danych z zestawu 'tips'

1. Tworzenie wykresu rozrzutu:
   - `sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='sex', sizes=(10,200), palette='viridis', style='sex', col='time', row='smoker')`
     - Tworzy wykres rozrzutu na podstawie danych z zestawu 'tips'.
     - Parametry:
       - `data`: Zestaw danych używany do wygenerowania wykresu.
       - `x`: Nazwa kolumny zawierającej dane dla osi x.
       - `y`: Nazwa kolumny zawierającej dane dla osi y.
       - `size`: Nazwa kolumny określającej rozmiar punktów.
       - `hue`: Nazwa kolumny określającej kategorie, które będą różniły się kolorem.
       - `sizes`: Zakres rozmiarów punktów na wykresie.
       - `palette`: Paleta kolorów do użycia na wykresie.
       - `style`: Nazwa kolumny określającej kategorie, które będą różniły się stylem punktów.
       - `col`: Nazwa kolumny, według której dane będą grupowane w kolumny (wiele wykresów w jednym wierszu).
       - `row`: Nazwa kolumny, według której dane będą grupowane w wiersze (wiele wykresów w jednej kolumnie).

2. Dodanie tytułu:
   - `plt.title('Wykres rozrzutu - Total Bill vs Tip')`: Dodaje tytuł wykresu.

3. Wyświetlenie wykresu:
   - `plt.show()`: Wyświetla wykres.

Zastosowanie:
- Wykres rozrzutu jest używany do wizualizacji relacji między dwiema lub więcej zmiennymi numerycznymi. Dodatkowo, dzięki możliwości grupowania danych na wielu wykresach (za pomocą parametrów `col` i `row`), umożliwia analizę tych relacji w kontekście różnych kategorii.
"""

import seaborn as sns
import matplotlib.pyplot as plt

# Ustawienie rozmiaru czcionki dla seaborn
sns.set(font_scale=1.2)

# Załadowanie danych
df = sns.load_dataset('tips')

# Tworzenie wykresu rozrzutu
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='sex', sizes=(10,200), palette='viridis', style='sex', col='time', row='smoker')



# Wyświetlenie wykresu
plt.show()
```

```python
"""
Notatka 1.3: Wykres rozrzutu czasu (timepoint) względem sygnału (signal) w badaniach fMRI

1. Wczytywanie danych:
   - `df = sns.load_dataset('fmri')`: Wczytuje zestaw danych fMRI z biblioteki Seaborn.

2. Podgląd pierwszych trzech wierszy danych:
   - `df.head(3)`: Wyświetla pierwsze trzy wiersze danych, aby uzyskać wgląd w ich strukturę i format.

3. Tworzenie wykresu rozrzutu (relplot):
   - `sns.relplot(data=df, x='timepoint', y='signal')`: Tworzy wykres rozrzutu czasu (timepoint) względem sygnału (signal) na podstawie danych z wczytanego zestawu.

Zastosowanie:
- Wykres rozrzutu czasu względem sygnału jest przydatny do wizualizacji zmian sygnału w czasie w badaniach fMRI. Pozwala zauważyć trendy, fluktuacje i inne wzorce w danych, które mogą być istotne dla analizy i interpretacji wyników badań.
"""
df = sns.load_dataset('fmri')
df.head(3)

sns.relplot(data=df, x='timepoint', y='signal')
```

```python
"""
Notatka 1.3: Wykres linii czasu (timepoint) względem sygnału (signal) w badaniach fMRI

1. Tworzenie wykresu linii (relplot):
   - `sns.relplot(data=df, x='timepoint', y='signal', kind='line')`: Tworzy wykres linii czasu (timepoint) względem sygnału (signal) na podstawie danych z wczytanego zestawu.
   - `kind='line'` określa typ wykresu jako liniowy, co oznacza, że punkty danych są połączone linią.
"""

sns.relplot(data=df, x='timepoint', y='signal', kind='line')
```

```python
"""
Notatka 1.3: Wykres linii czasu (timepoint) względem sygnału (signal) w badaniach fMRI

1. Tworzenie wykresu linii (relplot):
   - `sns.relplot(data=df, x='timepoint', y='signal', kind='line', ci=None)`: Tworzy wykres linii czasu (timepoint) względem sygnału (signal) na podstawie danych z wczytanego zestawu.
   -`kind='line'` określa typ wykresu jako liniowy, co oznacza, że punkty danych są połączone linią.
   -`ci=None` wyłącza przedziały ufności dla linii, co oznacza, że na wykresie nie będą wyświetlane cieniowane obszary wokół linii.
"""
sns.relplot(data=df, x='timepoint', y='signal', kind='line', ci=None)
```

```python
"""
Notatka 1.3: Tworzy wykres linii czasu (timepoint) względem sygnału (signal) na podstawie danych z wczytanego zestawu.
1. Tworzenie wykresu linii (relplot):
   - `sns.relplot(data=df, x='timepoint', y='signal', hue='event', kind='line', ci=None, style='event', markers=True)`:
   - `kind='line'`: Określa typ wykresu jako liniowy, co oznacza, że punkty danych są połączone linią.
   - `ci=None`: Wyłącza przedziały ufności dla linii, co oznacza, że na wykresie nie będą wyświetlane cieniowane obszary wokół linii.
   - `hue='event'`: Określa, że dane zostaną podzielone na grupy (event), a każda grupa będzie reprezentowana na wykresie różnym kolorem.
   - `style='event'`: Określa, że styl linii będzie zmieniany w zależności od grupy (event), co oznacza, że linie będą miały różne style w zależności od grupy.
   - `markers=True`: Włącza wyświetlanie punktów danych na wykresie, co oznacza, że punkty danych będą oznaczone punktami na wykresie linii.
"""
import seaborn as sns

# Tworzenie wykresu linii
sns.relplot(data=df, x='timepoint', y='signal', hue='event', kind='line', ci=None, style='event', markers=True)
```

```python
"""
Notatka 1.3: Tworzenie wykresu linii czasu (timepoint) względem sygnału (signal) na podstawie danych z wczytanego zestawu.
1. Tworzenie wykresu linii (relplot):
   - `sns.relplot(data=df, x='timepoint', y='signal', hue='event', kind='line', ci=None, style='event', markers=True, col='region')`:
   - `kind='line'`: Określa typ wykresu jako liniowy, co oznacza, że punkty danych są połączone linią.
   - `ci=None`: Wyłącza przedziały ufności dla linii, co oznacza, że na wykresie nie będą wyświetlane cieniowane obszary wokół linii.
   - `hue='event'`: Określa, że dane zostaną podzielone na grupy (event), a każda grupa będzie reprezentowana na wykresie różnym kolorem.
   - `style='event'`: Określa, że styl linii będzie zmieniany w zależności od grupy (event), co oznacza, że linie będą miały różne style w zależności od grupy.
   - `markers=True`: Włącza wyświetlanie punktów danych na wykresie, co oznacza, że punkty danych będą oznaczone punktami na wykresie linii.
   - `col='region'`: Tworzy osobne kolumny na wykresie dla każdego regionu, co pozwala na porównanie danych między różnymi regionami.
"""
import seaborn as sns

# Tworzenie wykresu linii
sns.relplot(data=df, x='timepoint', y='signal', hue='event', kind='line', ci=None, style='event', markers=True, col='region')

```

```python
"""
Notatka 1.4: Tworzenie wykresu linii czasu (timepoint) względem sygnału (signal) na podstawie danych z wczytanego zestawu.

1. Tworzenie wykresu linii (relplot):
   - `sns.relplot(data=df, x='timepoint', y='signal', hue='event', kind='line', ci=None, style='event', markers=True, col='subject', col_wrap=3)`:
     - `kind='line'`: Określa typ wykresu jako liniowy, co oznacza, że punkty danych są połączone linią.
     - `ci=None`: Wyłącza przedziały ufności dla linii, co oznacza, że na wykresie nie będą wyświetlane cieniowane obszary wokół linii.
     - `hue='event'`: Określa, że dane zostaną podzielone na grupy (event), a każda grupa będzie reprezentowana na wykresie różnym kolorem.
     - `style='event'`: Określa, że styl linii będzie zmieniany w zależności od grupy (event), co oznacza, że linie będą miały różne style w zależności od grupy.
     - `markers=True`: Włącza wyświetlanie punktów danych na wykresie, co oznacza, że punkty danych będą oznaczone punktami na wykresie linii.
     - `col='subject'`: Określa, że dane zostaną podzielone na kolumny (subject), a każda kolumna będzie reprezentowała innego podmiotu.
     - `col_wrap=3`: Określa, że maksymalnie 3 kolumny powinny być używane w układzie kolumnowym, zanim rozpocznie się nowy wiersz, co pomaga w kontroli układu, gdy mamy wiele kategorii w `col`.
"""

sns.relplot(data=df, x='timepoint', y='signal', hue='event', kind='line', ci=None, style='event', markers=True, col='subject', col_wrap=3)
```
