Notatka 1 Przygotowanie danych.

```python
"""
Notatka: Załadowanie i wstępne wyświetlenie danych z zestawu 'tips'

Poniżej znajduje się kod załadowania danych z zestawu 'tips' za pomocą biblioteki Seaborn oraz wyświetlenie trzech pierwszych wierszy z wczytanych danych.

1. Załadowanie danych:
   - `df = sns.load_dataset('tips')`: Wczytuje zestaw danych o napiwkach ('tips') za pomocą funkcji `load_dataset` z biblioteki Seaborn.

2. Wstępne wyświetlenie danych:
   - `df.head(3)`: Wyświetla trzy pierwsze wiersze z wczytanych danych, aby zapoznać się z ich strukturą i zawartością.
"""

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df = sns.load_dataset('tips')
df.head(3)
```

Notatka 2 Stworzenie wykresu regplot().

```python
"""
Notatka: Wykres regresji liniowej na podstawie danych z zestawu 'tips'

Poniżej znajduje się kod tworzący wykres regresji liniowej na podstawie danych z wczytanego zestawu 'tips' za pomocą biblioteki Seaborn.

1. Tworzenie wykresu regresji liniowej:
   - `sns.set()`: Ustawia domyślne ustawienia stylu i kolorów za pomocą funkcji `set` z biblioteki Seaborn.
   - `sns.regplot(data=df, x='total_bill', y='tip')`: Tworzy wykres regresji liniowej na podstawie danych z kolumn 'total_bill' (łączna kwota rachunku) i 'tip' (napiwek) z wczytanego zestawu danych.
"""


sns.set()
sns.regplot(data=df, x='total_bill', y='tip', color='red')
```
