```python
"""
Notatka: Załadowanie zestawu danych "titanic" za pomocą biblioteki Seaborn i wyświetlenie początkowych wierszy ramki danych

Ten skrypt zawiera kod do załadowania zestawu danych "titanic" za pomocą funkcji `load_dataset` z biblioteki Seaborn i wyświetlenia początkowych wierszy ramki danych.

1. Załadowanie zestawu danych "titanic":
   - `from seaborn import load_dataset`: Importuje funkcję `load_dataset` z modułu Seaborn.
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `df = load_dataset('titanic')`: Ładuje zestaw danych "titanic" za pomocą funkcji `load_dataset` z modułu Seaborn i przypisuje go do zmiennej `df`.
2. Wyświetlenie początkowych wierszy ramki danych:
   - `df.head()`: Wyświetla pierwsze pięć wierszy ramki danych `df`.
"""

import plotly.express as px
from seaborn import load_dataset

df = load_dataset('titanic')
df.head()

```

```python
"""
Notatka: Tworzenie wykresu skrzypcowego (violin plot) dla danych dotyczących wieku i przeżycia, kolorowanie według przeżycia

Ten skrypt zawiera kod do tworzenia wykresu skrzypcowego (violin plot) dla danych dotyczących wieku i przeżycia, kolorowanie według przeżycia.

1. Tworzenie wykresu skrzypcowego:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.violin(df, x='survived', y='age', color='survived', points='all', box=True)`: Tworzy wykres skrzypcowy na podstawie danych z ramki danych `df`. Wieku pasażera reprezentowany jest na osi Y, a przeżycia na osi X. Skrzypce są kolorowane według informacji o przeżyciu. Opcja `points='all'` oznacza, że wszystkie punkty danych zostaną wyświetlone na wykresie. Parametr `box=True` dodaje do wykresu również box plot.
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `x='survived'`: Nazwa kolumny reprezentującej wartości na osi X (przeżycie).
       - `y='age'`: Nazwa kolumny reprezentującej wartości na osi Y (wiek).
       - `color='survived'`: Kolumna, według której kolorowane są obszary skrzypiec (informacja o przeżyciu).
       - `points='all'`: Określa, że wszystkie punkty danych zostaną wyświetlone na wykresie.
"""

px.violin(df, x='survived', y='age', color='survived', points='all', box=True)
```

```python
"""
Notatka: Tworzenie wykresu skrzypcowego (violin plot) dla danych dotyczących wieku i przeżycia, kolorowanie według przeżycia, tryb nakładania skrzypiec

Ten skrypt zawiera kod do tworzenia wykresu skrzypcowego (violin plot) dla danych dotyczących wieku i przeżycia, kolorowanie według przeżycia, z użyciem trybu nakładania skrzypiec.

1. Tworzenie wykresu skrzypcowego:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.violin(df, y='age', color='survived', violinmode='overlay')`: Tworzy wykres skrzypcowy na podstawie danych z ramki danych `df`. Wieku pasażera reprezentowany jest na osi Y, a przeżycia są kolorowane według informacji o przeżyciu. Parametr `violinmode='overlay'` określa tryb nakładania skrzypiec, co oznacza, że skrzypce dla różnych kategorii nachodzą na siebie.
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `y='age'`: Nazwa kolumny reprezentującej wartości na osi Y (wiek).
       - `color='survived'`: Kolumna, według której kolorowane są obszary skrzypiec (informacja o przeżyciu).
       - `violinmode='overlay'`: Tryb nakładania skrzypiec, co oznacza, że skrzypce dla różnych kategorii nachodzą na siebie.
"""

px.violin(df,y='age', color='survived', violinmode='overlay')
```
