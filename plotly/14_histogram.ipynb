```python
"""
Notatka: Załadowanie zestawu danych "mpg" za pomocą biblioteki Seaborn i wyświetlenie początkowych wierszy ramki danych

Ten skrypt zawiera kod do załadowania zestawu danych "mpg" za pomocą funkcji `load_dataset` z biblioteki Seaborn i wyświetlenia początkowych wierszy ramki danych.

1. Załadowanie zestawu danych "mpg":
   - `from seaborn import load_dataset`: Importuje funkcję `load_dataset` z modułu Seaborn.
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `df = load_dataset('mpg')`: Ładuje zestaw danych "mpg" za pomocą funkcji `load_dataset` z modułu Seaborn i przypisuje go do zmiennej `df`.
2. Wyświetlenie początkowych wierszy ramki danych:
   - `df.head()`: Wyświetla pierwsze pięć wierszy ramki danych `df`.
"""

from seaborn import load_dataset
import plotly.express as px

df = load_dataset('mpg')
df.head()
```

```python
"""
Notatka: Tworzenie histogramu dla zmiennej "mpg" za pomocą biblioteki Plotly Express

Ten skrypt zawiera kod do tworzenia histogramu dla zmiennej "mpg" za pomocą funkcji `px.histogram()` z biblioteki Plotly Express.

1. Tworzenie histogramu:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.histogram(df, x='mpg', nbins=50, title='Distribution of variable mpg')`: Tworzy histogram na podstawie danych z ramki danych `df`, gdzie wartości na osi X są pobierane z kolumny `mpg`. Histogram zawiera 50 przedziałów (nbins) i ma tytuł "Distribution of variable mpg".
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `x='mpg'`: Nazwa kolumny reprezentującej wartości na osi X (zmienna "mpg").
       - `nbins=50`: Liczba przedziałów (binów), na które zostanie podzielony histogram.
       - `title='Distribution of variable mpg'`: Tytuł histogramu.
"""

px.histogram(df, x='mpg', nbins=50, title='Distribution of variable mpg')
```

```python
"""
Notatka: Tworzenie histogramu dla zmiennej "mpg" z znormalizowaną gęstością prawdopodobieństwa za pomocą biblioteki Plotly Express

Ten skrypt zawiera kod do tworzenia histogramu dla zmiennej "mpg" za pomocą funkcji `px.histogram()` z biblioteki Plotly Express, gdzie gęstość prawdopodobieństwa jest znormalizowana.

1. Tworzenie histogramu z znormalizowaną gęstością prawdopodobieństwa:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.histogram(df, x='mpg', nbins=50, title='Distribution of variable mpg', histnorm='probability density')`: Tworzy histogram na podstawie danych z ramki danych `df`, gdzie wartości na osi X są pobierane z kolumny `mpg`. Histogram zawiera 50 przedziałów (nbins), ma tytuł "Distribution of variable mpg" i gęstość prawdopodobieństwa jest znormalizowana.
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `x='mpg'`: Nazwa kolumny reprezentującej wartości na osi X (zmienna "mpg").
       - `nbins=50`: Liczba przedziałów (binów), na które zostanie podzielony histogram.
       - `title='Distribution of variable mpg'`: Tytuł histogramu.
       - `histnorm='probability density'`: Określa, że gęstość prawdopodobieństwa ma być znormalizowana.
"""

px.histogram(df, x='mpg', nbins=50, title='Distribution of variable mpg', histnorm='probability density')
```

```python
"""
Notatka: Tworzenie histogramów dla zmiennej "mpg" ze znormalizowaną gęstością prawdopodobieństwa dla różnych kategorii za pomocą biblioteki Plotly Express

Ten skrypt zawiera kod do tworzenia histogramów dla zmiennej "mpg" za pomocą funkcji `px.histogram()` z biblioteki Plotly Express, gdzie gęstość prawdopodobieństwa jest znormalizowana dla różnych kategorii.

1. Tworzenie histogramów dla różnych kategorii z znormalizowaną gęstością prawdopodobieństwa:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.histogram(df, x='mpg', nbins=50, title='Distribution of variable mpg', histnorm='probability density', facet_row='origin', color='origin')`: Tworzy histogramy na podstawie danych z ramki danych `df`, gdzie wartości na osi X są pobierane z kolumny `mpg`. Histogramy zawierają 50 przedziałów (nbins), mają tytuł "Distribution of variable mpg", gęstość prawdopodobieństwa jest znormalizowana, a histogramy są tworzone dla różnych kategorii określonych przez kolumnę `origin`. Histogramy są dodatkowo kolorowane na podstawie kategorii.
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `x='mpg'`: Nazwa kolumny reprezentującej wartości na osi X (zmienna "mpg").
       - `nbins=50`: Liczba przedziałów (binów), na które zostanie podzielony każdy histogram.
       - `title='Distribution of variable mpg'`: Tytuł histogramów.
       - `histnorm='probability density'`: Określa, że gęstość prawdopodobieństwa ma być znormalizowana.
       - `facet_row='origin'`: Tworzy osobne wiersze histogramów dla każdej kategorii zdefiniowanej w kolumnie `origin`.
       - `color='origin'`: Koloruje histogramy na podstawie kategorii zdefiniowanych w kolumnie `origin`.

"""

px.histogram(df, x='mpg', nbins=50, title='Distribution of variable mpg', histnorm='probability density',
             facet_row='origin', color='origin')
```

```python
"""
Notatka: Tworzenie histogramu dla zmiennej "mpg" dla pojazdów pochodzących z USA i Europy za pomocą biblioteki Plotly Express

Ten skrypt zawiera kod do tworzenia histogramu dla zmiennej "mpg" dla pojazdów pochodzących z USA i Europy za pomocą funkcji `px.histogram()` z biblioteki Plotly Express.

1. Tworzenie histogramu dla pojazdów pochodzących z USA i Europy:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `dff = df[(df.origin == 'usa') | (df.origin == 'europe')]`: Tworzy nową ramkę danych `dff`, która zawiera tylko dane dla pojazdów pochodzących z USA lub Europy.
   - `px.histogram(dff, x='mpg', color='origin', marginal='rug', title='Distribution of variable mpg (usa vs. europe)', opacity=0.7)`: Tworzy histogram na podstawie danych z ramki danych `dff`, gdzie wartości na osi X są pobierane z kolumny `mpg`. Histogram jest kolorowany na podstawie kolumny `origin`, posiada dodatkowe oznaczenia marginesowe w postaci 'rug', ma tytuł "Distribution of variable mpg (usa vs. europe)" i ustawia przezroczystość na 0.7.
     - Parametry:
       - `dff`: Ramka danych zawierająca dane do wykresu, zawierająca tylko dane dla pojazdów pochodzących z USA lub Europy.
       - `x='mpg'`: Nazwa kolumny reprezentującej wartości na osi X (zmienna "mpg").
       - `color='origin'`: Kolumna, według której kolorowane są słupki histogramu (pochodzenie pojazdu).
       - `marginal='rug'`: Dodaje oznaczenia marginesowe w postaci 'rug' na osi X i Y.
       - `title='Distribution of variable mpg (usa vs. europe)'`: Tytuł histogramu.
       - `opacity=0.7`: Określa przezroczystość histogramu.
"""


dff = df[(df.origin == 'usa') | (df.origin == 'europe')]
px.histogram(dff, x='mpg', color='origin', marginal='rug', title='Distribution of variable mpg (usa vs. europe)', opacity=0.7) #marginal: etc. box, violin
```

```python
"""
Notatka: Tworzenie wykresu rozkładu dla zmiennej "mpg" za pomocą biblioteki Plotly Figure Factory

Ten skrypt zawiera kod do tworzenia wykresu rozkładu dla zmiennej "mpg" za pomocą funkcji `ff.create_distplot()` z biblioteki Plotly Figure Factory.

1. Tworzenie wykresu rozkładu dla zmiennej "mpg":
   - `import plotly.figure_factory as ff`: Importuje moduł Plotly Figure Factory jako `ff`.
   - `fig = ff.create_distplot([df.mpg], group_labels=['mpg'])`: Tworzy wykres rozkładu na podstawie danych z kolumny `mpg` ramki danych `df`. Wykres zawiera tylko jedną grupę danych, oznaczoną jako 'mpg'.
     - Parametry:
       - `[df.mpg]`: Lista zawierająca dane do wykresu rozkładu.
       - `group_labels=['mpg']`: Etykieta grupy danych na wykresie rozkładu.
"""

import plotly.figure_factory as ff

fig = ff.create_distplot([df.mpg], group_labels=['mpg'])
fig.show()
```

```python
"""
Notatka: Tworzenie wykresu rozkładu dla zmiennych "mpg" i "acceleration" za pomocą biblioteki Plotly Figure Factory

Ten skrypt zawiera kod do tworzenia wykresu rozkładu dla zmiennych "mpg" i "acceleration" za pomocą funkcji `ff.create_distplot()` z biblioteki Plotly Figure Factory.

1. Tworzenie wykresu rozkładu dla zmiennych "mpg" i "acceleration":
   - `import plotly.figure_factory as ff`: Importuje moduł Plotly Figure Factory jako `ff`.
   - `fig = ff.create_distplot([df.mpg, df.acceleration], group_labels=['mpg', 'Acceleration'])`: Tworzy wykres rozkładu na podstawie danych z kolumn "mpg" i "acceleration" z ramki danych `df`. Wykres zawiera dwie grupy danych: "mpg" i "Acceleration".
     - Parametry:
       - `[df.mpg, df.acceleration]`: Lista zawierająca dane dla obu zmiennych do wykresu rozkładu.
       - `group_labels=['mpg', 'Acceleration']`: Etykiety grup danych na wykresie rozkładu.
"""

import plotly.figure_factory as ff

fig = ff.create_distplot([df.mpg, df.acceleration], group_labels=['mpg', 'Acceleration'])
fig.show()
```

```python
"""
Notatka: Tworzenie wykresu rozkładu dla dwóch rozkładów normalnych za pomocą biblioteki Plotly Figure Factory

Ten skrypt zawiera kod do tworzenia wykresu rozkładu dla dwóch rozkładów normalnych za pomocą funkcji `ff.create_distplot()` z biblioteki Plotly Figure Factory.

1. Tworzenie wykresu rozkładu dla dwóch rozkładów normalnych:
   - `import numpy as np`: Importuje moduł numpy jako `np`.
   - `x1 = np.random.randn(1100)`: Generuje 1100 losowych liczb z rozkładu normalnego N(0,1) i przypisuje je do zmiennej `x1`.
   - `x2 = np.random.randn(1100) + 2`: Generuje 1100 losowych liczb z rozkładu normalnego N(2,1) i przypisuje je do zmiennej `x2`.
   - `fig = ff.create_distplot([x1, x2],['Rozkład normalny N(0,1)', 'Rozkład normalny N(2,1)'], curve_type='normal', bin_size=0.2, colors=['slategray', 'magenta'])`: Tworzy wykres rozkładu na podstawie danych z zmiennych `x1` i `x2`, opatrując je odpowiednimi etykietami. Wykres zawiera dwa rozkłady normalne opisane etykietami 'Rozkład normalny N(0,1)' i 'Rozkład normalny N(2,1)'. Dla wykresu używany jest typ krzywej 'normal', a wielkość przedziałów (binów) ustawiona jest na 0.2. Dodatkowo, kolory krzywych są ustawione na 'slategray' i 'magenta'.
     - Parametry:
       - `[x1, x2]`: Lista zawierająca dane dla dwóch rozkładów normalnych.
       - `['Rozkład normalny N(0,1)', 'Rozkład normalny N(2,1)']`: Lista zawierająca etykiety dla obu rozkładów.
       - `curve_type='normal'`: Określa typ krzywej na "normal", co oznacza, że wykres będzie przedstawiał prawdziwe rozmieszczenie danych.
       - `bin_size=0.2`: Określa wielkość przedziałów (binów) na wykresie.
       - `colors=['slategray', 'magenta']`: Lista zawierająca kolory dla poszczególnych rozkładów.
"""

import numpy as np

x1 = np.random.randn(1100)
x2 = np.random.randn(1100) + 2

fig = ff.create_distplot([x1,x2],['Rozkład normalny N(2,1)','Rozkład normalny N(0,1)'],
                         curve_type='normal',
                         bin_size=0.2,
                         colors=['slategray', 'magenta'])
fig.show()
```
