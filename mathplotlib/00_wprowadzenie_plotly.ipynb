Notatka 1: Importowanie bibliotek.

```python
"""
Notatka 1: Importowanie bibliotek do analizy danych

Ten skrypt zawiera importowanie niezbędnych bibliotek do analizy i wizualizacji danych w języku Python.

Biblioteki:
1. Matplotlib: Biblioteka służąca do tworzenia statycznych wykresów.
2. Seaborn: Biblioteka rozszerzająca możliwości Matplotlib o bardziej zaawansowane funkcje wizualizacji.
3. Pandas: Biblioteka do manipulacji i analizy danych.
4. NumPy: Biblioteka do obliczeń numerycznych w Pythonie.
"""

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

Notatka 2: Przygotowanie danych.

```python
"""
Notatka 2: Ustawienie opcji drukowania i generowanie danych

Ten skrypt zawiera ustawienie opcji drukowania i generowanie danych do analizy.

1. Ustawienie opcji drukowania:
   - Funkcja `np.set_printoptions(suppress=True, precision=8)` ustawia opcje drukowania dla tablic NumPy.
   - `suppress=True` powoduje, że NumPy pomija zapisywanie bardzo małych wartości w notacji naukowej.
   - `precision=8` określa dokładność wartości dziesiętnych do 8 miejsc po przecinku.

2. Generowanie danych:
   - `x = np.linspace(0,20,100)` generuje 100 równomiernie rozłożonych liczb od 0 do 20.
   - `y = np.random.randn(100,4)` generuje macierz 100x4 liczb z rozkładu normalnego.
   - `y = np.cumsum(y, axis=0)` oblicza skumulowaną sumę dla wygenerowanych danych wzdłuż pierwszej osi.

Te dane mogą być używane do przeprowadzenia różnych analiz i eksperymentów.
"""

import numpy as np

# Ustawienie opcji drukowania
np.set_printoptions(suppress=True, precision=8)

# Generowanie danych
x = np.linspace(0, 20, 100)
y = np.random.randn(100, 4)
y = np.cumsum(y, axis=0)
```

Notatke 3.1: Wykresy i ich zastosowanie.

```python
"""
Notatka 3.1: Rodzaje wykresów w bibliotece Plotly i ich zastosowania

Biblioteka Plotly oferuje szeroki zakres wykresów interaktywnych, które mogą być wykorzystywane w różnych scenariuszach analizy danych i wizualizacji. Poniżej przedstawiono kilka popularnych rodzajów wykresów dostępnych w Plotly oraz ich główne zastosowania:

1. Wykres liniowy (plot lub scatter):
   - Zastosowanie: Wykorzystywany do pokazywania zmiany wartości zmiennej w czasie lub w zależności od innej zmiennej. Idealny do analizy trendów i porównań.

2. Wykres punktowy (scatter):
   - Zastosowanie: Służy do przedstawiania punktów danych na płaszczyźnie, gdzie każdy punkt może być różny pod względem koloru, rozmiaru lub kształtu. Doskonały do analizy zależności między dwiema zmiennymi.

3. Wykres słupkowy (bar):
   - Zastosowanie: Nadaje się do porównywania wielkości lub ilości między różnymi kategoriami. Może być wykorzystany do prezentacji danych kategorycznych.

4. Wykres kołowy (pie):
   - Zastosowanie: Idealny do pokazywania udziałów procentowych poszczególnych kategorii w całkowitej wartości. Najczęściej używany, gdy chcesz podkreślić udziały procentowe w całej populacji.

5. Histogram (histogram):
   - Zastosowanie: Wykorzystywany do analizy rozkładu danych numerycznych poprzez pokazanie częstotliwości występowania wartości w określonych przedziałach.

6. Wykres pudełkowy (box lub boxplot):
   - Zastosowanie: Służy do przedstawiania rozkładu danych numerycznych wraz z wykrywaniem wartości odstających, mediany, kwartylów itp. Przydatny do analizy rozkładu danych i porównań między grupami.

7. Wykres powierzchniowy (surface):
   - Zastosowanie: Nadaje się do wizualizacji danych trójwymiarowych na powierzchniach. Często używany w analizie danych przestrzennych lub modelowaniu matematycznym.

8. Wykres heatmap (heatmap):
   - Zastosowanie: Idealny do przedstawiania danych w formie kolorowej mapy, gdzie intensywność kolorów reprezentuje wartości numeryczne w określonych punktach.

Wybór odpowiedniego rodzaju wykresu zależy od rodzaju danych, które chcesz przedstawić, oraz celu analizy. Dobrze dobrane wykresy mogą pomóc w efektywnej komunikacji informacji i lepszym zrozumieniu danych przez odbiorcę.
"""
```

Notatka 3.2: Stworzenie wykresu liniowego plt.plot().

```python
"""
Notatka 3.2: Tworzenie wykresu liniowego za pomocą plt.plot()

Ten skrypt zawiera kod do tworzenia wykresu liniowego za pomocą funkcji plt.plot().

1. Tworzenie wykresu liniowego:
   - `plt.plot(x, y)` tworzy wykres liniowy na podstawie danych x i y.
   - `x` to wektor zawierający wartości na osi x.
   - `y` to macierz zawierająca wartości na osi y, skumulowane wcześniej.
   - Funkcja `plt.plot()` rysuje linię, łącząc punkty danych ze sobą.

Ten wykres liniowy może być używany do wizualizacji trendów lub zmian wartości w czasie.
"""

import matplotlib.pyplot as plt

# Tworzenie wykresu liniowego
plt.plot(x, y)

# Wyświetlenie wykresu
plt.show()

```

Notatka 3.3: Stworzenie wykresu liniowego z dodatkowymi parametrami.

```python
"""
Notatka 3.3: Tworzenie wykresu liniowego z różnymi parametrami

Ten skrypt zawiera kod do tworzenia wykresu liniowego za pomocą funkcji plt.plot() z różnymi parametrami.

1. Tworzenie wykresu liniowego:
   - `plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='o', markersize=8, alpha=0.7, label='sin(x)')`
     - Tworzy wykres liniowy na podstawie danych x i y.
     - Parametry:
       - `color='red'`: Kolor linii ustawiony na czerwony. (string: 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w' lub nazwa koloru)
       - `linestyle='--'`: Styl linii ustawiony na przerywaną. ('-', '--', '-.', ':', 'None', ' ', '', itp.)
       - `linewidth=2`: Grubość linii ustawiona na 2. (float)
       - `marker='o'`: Znak punktu ustawiony na kółko. ('.', ',', 'o', 'v', '^', '<', '>', '1', '2', '3', '4', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_', itp.)
       - `markersize=8`: Rozmiar znaczników ustawiony na 8. (float)
       - `alpha=0.7`: Poziom przezroczystości linii ustawiony na 0.7. (float, od 0 do 1)
       - `label='sin(x)'`: Etykieta serii danych ustawiona na 'sin(x)'. (string)

2. Dodanie tytułu i etykiet osi:
   - `plt.title('Wykres funkcji sin(x)')`: Dodaje tytuł wykresu. (string)
   - `plt.xlabel('x')`: Dodaje etykietę osi x. (string)
   - `plt.ylabel('sin(x)')`: Dodaje etykietę osi y. (string)

3. Dodanie legendy:
   - `plt.legend()`: Dodaje legendę wykresu.

4. Wyświetlenie wykresu:
   - `plt.show()`: Wyświetla wykres.
"""

import matplotlib.pyplot as plt
import numpy as np

# Dane do wykresu
x = np.linspace(0, 20, 200)
y = np.sin(x)

# Tworzenie wykresu z różnymi parametrami
plt.plot(x, y, color='red', linewidth=2, marker='o', markersize=8, alpha=0.7, label='sin(x)')
plt.plot(x, y, color='blue', linestyle='--', linewidth=2.5, markersize=8, alpha=0.7, label='sin(x)')
# Dodanie tytułu i etykiet osi
plt.title('Wykres funkcji sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Dodanie legendy
plt.legend()

# Wyświetlenie wykresu
plt.show()
```

Notatka 3.4 Stworzenie wykresu punktowego plt.scatter().



```python
"""
Notatka 3.4: Wykres punktowy na podstawie danych z zestawu 'tips'

1. Tworzenie wykresu punktowego:
   - `plt.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, alpha=None, linewidths=None, edgecolors=None, label=None)`
     - Tworzy wykres punktowy na podstawie danych x i y.
     - Parametry:
       - `x`: Wartości na osi x.
       - `y`: Wartości na osi y.
       - `s` (opcjonalny): Rozmiar punktów. Może być pojedynczą liczbą lub tablicą.
       - `c` (opcjonalny): Kolor punktów. Może być pojedynczą wartością lub tablicą. (string: 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w' lub nazwa koloru)
       - `marker` (opcjonalny): Symbol punktów. Może być jednym z: 'o', '^', 's', itp.
       - `cmap` (opcjonalny): Mapa kolorów do użycia dla kolorów punktów. Przykładowe predefiniowane mapy kolorów to: 'viridis', 'plasma', 'inferno', 'magma', 'cividis', 'Greys', 'Purples', 'Blues',
       - `alpha` (opcjonalny): Przezroczystość punktów. Określa poziom przezroczystości punktów na wykresie od 0 do 1.
       - `linewidths` (opcjonalny): Grubość krawędzi punktów. Określa grubość krawędzi punktów na wykresie.
       - `edgecolors` (opcjonalny): Kolor krawędzi punktów. Określa kolor krawędzi punktów na wykresie.
       - `label` (opcjonalny): Etykieta serii danych. Używany do określenia etykiety dla punktów na wykresie, co jest przydatne przy tworzeniu legendy.


2. Dodanie tytułu i etykiet osi:
   - `plt.title('Tytuł wykresu')`: Dodaje tytuł wykresu.
   - `plt.xlabel('Etykieta osi x')`: Dodaje etykietę osi x.
   - `plt.ylabel('Etykieta osi y')`: Dodaje etykietę osi y.

3. Dodanie legendy:
   - `plt.legend()`: Dodaje legendę wykresu.

4. Wyświetlenie wykresu:
   - `plt.show()`: Wyświetla wykres.

Zastosowanie:
- Wykres punktowy jest przydatny do zobrazowania zależności między dwiema zmiennymi oraz do identyfikacji wzorców lub tendencji w danych.

"""

import seaborn as sns
import matplotlib.pyplot as plt

# Załadowanie danych
df = sns.load_dataset('tips')
print(df.head())
print()
# Wykres punktowy
plt.scatter(x='total_bill', y='tip', data=df, s=25, c='red', marker='o', cmap='Blues', alpha=0.7, linewidths=1, edgecolor='blue')


# Dodanie tytułu i etykiet osi
plt.title('Wykres punktowy - Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# Wyświetlenie wykresu
plt.show()
```

Notatka 3.5 Stworzenie wykresu słupkowego plt.bar().

```python
"""
Notatka 3.5: Wykres słupkowy na podstawie danych z zestawu 'tips'

1. Tworzenie wykresu słupkowego:
   - `plt.bar(x, height, width=0.8, bottom=None, align='center', data=None, color=None, alpha=None, edgecolor=None, linewidth=None)`
     - Tworzy wykres słupkowy na podstawie danych x i wysokości słupków.
     - Parametry:
       - `x`: Wartości na osi x.
       - `height`: Wysokości słupków.
       - `width` (opcjonalny): Szerokość słupków. (float)
       - `bottom` (opcjonalny): Wysokość słupków bazowych. (float)
       - `align` (opcjonalny): Wyrównanie słupków. ('center', 'edge')
       - `data` (opcjonalny): Zestaw danych do użycia. (DataFrame)
       - `color` (opcjonalny): Kolor słupków. ('blue', 'red', 'green', 'orange', itp.)
       - `alpha` (opcjonalny): Przezroczystość słupków (od 0 do 1). (float)
       - `edgecolor` (opcjonalny): Kolor krawędzi słupków. ('black', 'gray', 'purple', itp.)
       - `linewidth` (opcjonalny): Grubość krawędzi słupków. (float)

2. Dodanie tytułu i etykiet osi:
   - `plt.title('Tytuł wykresu')`: Dodaje tytuł wykresu. (string)
   - `plt.xlabel('Etykieta osi x')`: Dodaje etykietę osi x. (string)
   - `plt.ylabel('Etykieta osi y')`: Dodaje etykietę osi y. (string)

3. Wyświetlenie wykresu:
   - `plt.show()`: Wyświetla wykres.

Zastosowanie:
- Wykres słupkowy jest używany do porównywania wielkości kategorii na osi x poprzez wysokość słupków na osi y.
"""

import seaborn as sns
import matplotlib.pyplot as plt

# Załadowanie danych
df = sns.load_dataset('tips')

# Wykres słupkowy z dostosowanymi parametrami
plt.bar(x='day', height='total_bill', data=df, color='blue', alpha=0.7, edgecolor='black', linewidth=2)

# Dodanie tytułu i etykiet osi
plt.title('Wykres słupkowy - Dzień vs Kwota rachunku')
plt.xlabel('Dzień')
plt.ylabel('Kwota rachunku')

# Wyświetlenie wykresu
plt.show()
```

Notatka 3.6 Stworzenie wykresu słupkowego plt.pie().

```python
"""
Notatka 3.6: Wykres kołowy na podstawie danych z zestawu 'tips'

1. Tworzenie wykresu kołowego:
   - `plt.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, startangle=None, radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, normalize=None)`
     - Tworzy wykres kołowy na podstawie danych x.
     - Parametry:
       - `x`: Wartości do wyświetlenia na wykresie kołowym.
       - `explode` (opcjonalny): Rozdzielność poszczególnych kawałków.
       - `labels` (opcjonalny): Etykiety dla poszczególnych kawałków.
       - `colors` (opcjonalny): Kolory dla poszczególnych kawałków.
       - `autopct` (opcjonalny): Format etykiet procentowych.
       - `pctdistance` (opcjonalny): Odległość od środka wykresu dla etykiet procentowych.
       - `shadow` (opcjonalny): Czy dodać efekt cienia do wykresu.
       - `startangle` (opcjonalny): Kąt startowy dla pierwszego kawałka.
       - `radius` (opcjonalny): Promień koła.
       - `counterclock` (opcjonalny): Kierunek rysowania kawałków, zgodnie z ruchem wskazówek zegara lub przeciwnie.
       - `wedgeprops` (opcjonalny): Właściwości kształtu kawałków.
       - `textprops` (opcjonalny): Właściwości etykiet tekstowych.
       - `center` (opcjonalny): Współrzędne środka koła.
       - `frame` (opcjonalny): Czy dodać ramkę do wykresu.
       - `rotatelabels` (opcjonalny): Czy obracać etykiety.
       - `normalize` (opcjonalny): Czy znormalizować wartości przed rysowaniem wykresu.

2. Dodanie tytułu:
   - `plt.title('Tytuł wykresu')`: Dodaje tytuł wykresu.

3. Wyświetlenie wykresu:
   - `plt.show()`: Wyświetla wykres.

Zastosowanie:
- Wykres kołowy jest używany do przedstawiania proporcji poszczególnych kategorii w stosunku do całości.
"""

import seaborn as sns
import matplotlib.pyplot as plt

# Załadowanie danych
df = sns.load_dataset('tips')

# Wykres kołowy
plt.pie(df['size'].value_counts(), labels=df['size'].value_counts().index, autopct='%1.1f%%', startangle=90)

# Dodanie tytułu
plt.title('Wykres kołowy - Rozmiar grupy')

# Wyświetlenie wykresu
plt.show()
```

Notatka 3.7 Stworzenie histogramu plt.hist().

```python
"""
Notatka 3.7: Histogram na podstawie danych z zestawu 'tips'

1. Tworzenie histogramu:
   - `plt.hist(x, bins=10, color=None, edgecolor=None, ...)`:
     - Tworzy histogram na podstawie danych x.
     - Parametry:
       - `x`: Dane numeryczne do utworzenia histogramu.
       - `bins` (opcjonalny): Liczba lub lista przedziałów.
       - `color` (opcjonalny): Kolor słupków histogramu.
       - `edgecolor` (opcjonalny): Kolor krawędzi słupków histogramu.
       - Inne opcjonalne parametry: alpha, density, weights, range, histtype itp.

2. Dodanie tytułu i etykiet osi:
   - `plt.title('Tytuł wykresu')`: Dodaje tytuł wykresu.
   - `plt.xlabel('Etykieta osi x')`: Dodaje etykietę osi x.
   - `plt.ylabel('Etykieta osi y')`: Dodaje etykietę osi y.

3. Wyświetlenie wykresu:
   - `plt.show()`: Wyświetla wykres.
"""

import seaborn as sns
import matplotlib.pyplot as plt

# Załadowanie danych
df = sns.load_dataset('tips')

# Tworzenie histogramu
plt.hist(df['total_bill'], bins=20, color='skyblue', edgecolor='black')

# Dodanie tytułu
plt.title('Histogram - Rozkład wartości całkowitych rachunków')

# Dodanie etykiet osi
plt.xlabel('Wartość całkowita rachunku')
plt.ylabel('Częstotliwość')

# Wyświetlenie wykresu
plt.show()
```
