```python
"""
Notatka: Importowanie biblioteki Plotly Express oraz wczytywanie danych zestawu Iris

Ten skrypt zawiera kod do importowania biblioteki Plotly Express oraz wczytywania danych zestawu danych Iris.
"""

import plotly.express as px

data = px.data.iris()
data.head()

```

```python
"""
Notatka: Tworzenie macierzy wykresów punktowych za pomocą funkcji px.scatter_matrix()

Ten skrypt zawiera kod do tworzenia macierzy wykresów punktowych za pomocą funkcji `px.scatter_matrix()` z różnymi parametrami.

1. Tworzenie macierzy wykresów punktowych:
   - `px.scatter_matrix(data_frame=data, dimensions=['sepal_length','sepal_width','petal_length','petal_width'], color='species', height=800, title='Scatter Matrix')`
     - Tworzy macierz wykresów punktowych na podstawie danych z ramki danych `data`. Wykresy w macierzy pokazują zależności między wybranymi wymiarami (kolumnami).
     - Parametry:
       - `data_frame=data`: Ramka danych zawierająca dane do wykresu.
       - `dimensions=['sepal_length','sepal_width','petal_length','petal_width']`: Lista wymiarów (kolumn), które mają być uwzględnione w macierzy wykresów.
       - `color='species'`: Kolorowanie punktów na wykresach na podstawie wartości w kolumnie `species`.
       - `height=800`: Wysokość macierzy wykresów w pikselach.
       - `title='Scatter Matrix'`: Tytuł macierzy wykresów.
"""

px.scatter_matrix(data_frame=data,
                  dimensions=['sepal_length','sepal_width','petal_length','petal_width'],
                  color='species',
                  height=800,
                  title='Scatter Matrix')
```
