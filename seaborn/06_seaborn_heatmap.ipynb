```python
"""
Notatka: Tworzenie heatmapy na podstawie losowych danych

1.Tworzenie losowych danych:

uni = np.random.randn(100, 100): Generuje losowe dane numeryczne o rozkadzie normalnym (Gaussa) o wymiarach 100x100 przy uyciu biblioteki NumPy.
Tworzenie heatmapy:

2. sns.heatmap(uni): Tworzy heatmap na podstawie wygenerowanych danych. Kada komrka heatmapy reprezentuje warto danych numerycznych, gdzie intensywno koloru wskazuje na warto liczbow.
Im janiejszy kolor, tym wiksza warto.

Zastosowanie:

Heatmapa jest uywana do wizualizacji danych w postaci kolorowej mapy, gdzie intensywno kolorw reprezentuje wartoci numeryczne w okrelonych punktach.
Jest przydatna do analizy zalenoci midzy dwoma zestawami danych oraz do identyfikacji wzorcw lub trendw.
"""
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

uni = np.random.randn(100, 100)
sns.heatmap(uni)
```

```python
"""
Notatka: Wczytanie danych i wstpne wywietlenie

Poniej znajduje si kod wczytujcy dane z zestawu 'flights' za pomoc biblioteki Seaborn oraz wywietlajcy kilka pierwszych wierszy z wczytanych danych.

1. Wczytanie danych:
   - `df = sns.load_dataset('flights')`: Wczytuje zestaw danych o lotach ('flights') za pomoc funkcji `load_dataset` z biblioteki Seaborn.

2. Wstpne wywietlenie danych:
   - `df.head()`: Wywietla kilka pierwszych wierszy z wczytanych danych, aby zapozna si z ich struktur i zawartoci.

Wynikiem dziaania kodu jest wywietlenie kilku pierwszych wierszy danych z zestawu 'flights'.
"""

df = sns.load_dataset('flights')
df.head()
```

```python
"""
Notatka: Tworzenie heatmapy na podstawie przeksztaconych danych

Heatmapa jest uywana do wizualizacji danych w postaci kolorowej mapy, gdzie intensywno kolorw reprezentuje wartoci numeryczne w okrelonych punktach. Jest przydatna do analizy zalenoci midzy dwoma zestawami danych oraz do identyfikacji wzorcw lub trendw.

1. Przeksztacenie danych:
   - `df = df.pivot(index='month', columns='year', values='passengers')`: Przeksztaca pierwotny zbir danych w form odpowiedni do wygenerowania heatmapy. Dane zostaj przeksztacone w posta tabeli przestawnej, gdzie indeksem jest miesic, kolumnami s lata, a wartociami s liczba pasaerw.

2. Tworzenie heatmapy:
   - `sns.heatmap(df)`: Tworzy heatmap na podstawie przeksztaconych danych. Kada komrka heatmapy reprezentuje warto danych numerycznych, gdzie intensywno koloru wskazuje na warto liczbow. Im janiejszy kolor, tym wiksza warto.

Zastosowanie:
-Korelacja midzy zmiennymi,
Analiza przestrzeni cech.
"""

df = df.pivot(index='month', columns='year', values='passengers')
sns.heatmap(df)
```

```python
"""
Notatka: Generowanie heatmapy na podstawie danych

Poniej znajduje si kod tworzcy heatmap na podstawie danych z ramki danych `df` za pomoc funkcji `heatmap` z biblioteki Seaborn.

1. Generowanie heatmapy:
   - `sns.heatmap(df, annot=True, fmt='d', cmap='viridis')`: Tworzy heatmap na podstawie danych z ramki danych `df`.
     - `annot=True`: Wcza wywietlanie wartoci liczbowych w kadej komrce heatmapy.
     - `fmt='d'`: Okrela format wywietlanych liczb jako liczby cakowite.
     - `cmap='viridis'`: Okrela map kolorw dla heatmapy, w tym przypadku 'viridis'.

Wynikiem dziaania kodu jest wygenerowanie heatmapy na podstawie danych z ramki danych `df` z wywietlonymi wartociami liczbowymi w kadej komrce.


"""

sns.heatmap(df,annot=True, fmt='d', cmap='viridis')
```
