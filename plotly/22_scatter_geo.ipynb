```python
"""
Notatka: Wczytywanie danych dotyczących Gapminder

Ten skrypt zawiera kod do wczytywania danych dotyczących Gapminder za pomocą modułu `plotly.express`.

1. Wczytywanie danych:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `df = px.data.gapminder()`: Wczytuje dane dotyczące Gapminder za pomocą wbudowanej funkcji `gapminder()` z modułu Plotly Express i zapisuje je do zmiennej `df`.
   - `df.head()`: Wyświetla pierwsze pięć wierszy wczytanych danych w celu weryfikacji poprawności wczytania.
"""

import plotly.express as px

df = px.data.gapminder()
df.head()
```

```python
df_2007 = df[df.year == 2007]
df_2007.describe()
```

```python
"""
Notatka: Tworzenie interaktywnego wykresu punktowego na mapie świata

Ten skrypt zawiera kod do tworzenia interaktywnego wykresu punktowego na mapie świata za pomocą modułu `plotly.express`.

1. Tworzenie interaktywnego wykresu punktowego:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.scatter_geo(`: Tworzy interaktywny wykres punktowy na mapie świata za pomocą funkcji `scatter_geo` z modułu Plotly Express.
     - `df_2007`: Ramka danych zawierająca dane geograficzne.
     - `locations='iso_alpha'`: Kolumna zawierająca kody ISO dla poszczególnych krajów.
     - `color='continent'`: Kolumna, według której punkty na mapie są kolorowane (np. kontynent).
     - `hover_name='country'`: Nazwa kolumny wyświetlana w interaktywnym podglądzie punktów na mapie (np. nazwa kraju).
     - `size='pop'`: Kolumna zawierająca dane dotyczące wielkości punktów na mapie (np. populacja).
2. Wyświetlenie wykresu:
   - Wywołanie tej funkcji automatycznie wyświetli interaktywny wykres.
"""

px.scatter_geo(df_2007, locations='iso_alpha', color='continent', hover_name='country',
              size='pop')
```

```python
"""
Notatka: Tworzenie interaktywnego wykresu punktowego na mapie świata z określoną projekcją

Ten skrypt zawiera kod do tworzenia interaktywnego wykresu punktowego na mapie świata z określoną projekcją za pomocą modułu `plotly.express`.

1. Tworzenie interaktywnego wykresu punktowego:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.scatter_geo(`: Tworzy interaktywny wykres punktowy na mapie świata za pomocą funkcji `scatter_geo` z modułu Plotly Express.
     - `df_2007`: Ramka danych zawierająca dane geograficzne.
     - `locations='iso_alpha'`: Kolumna zawierająca kody ISO dla poszczególnych krajów.
     - `color='continent'`: Kolumna, według której punkty na mapie są kolorowane (np. kontynent).
     - `hover_name='country'`: Nazwa kolumny wyświetlana w interaktywnym podglądzie punktów na mapie (np. nazwa kraju).
     - `size='pop'`: Kolumna zawierająca dane dotyczące wielkości punktów na mapie (np. populacja).
     - `projection='natural earth'`: Określa projekcję mapy, w tym przypadku 'natural earth'.
2. Wyświetlenie wykresu:
   - Wywołanie tej funkcji automatycznie wyświetli interaktywny wykres.
"""

px.scatter_geo(df_2007, locations='iso_alpha', color='continent', hover_name='country',
              size='pop', projection='natural earth')
```

```python
"""
Notatka: Tworzenie interaktywnego wykresu punktowego na mapie świata z animacją

Ten skrypt zawiera kod do tworzenia interaktywnego wykresu punktowego na mapie świata z animacją za pomocą modułu `plotly.express`.

1. Tworzenie interaktywnego wykresu punktowego:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.scatter_geo(`: Tworzy interaktywny wykres punktowy na mapie świata za pomocą funkcji `scatter_geo` z modułu Plotly Express.
     - `df`: Ramka danych zawierająca dane geograficzne.
     - `locations='iso_alpha'`: Kolumna zawierająca kody ISO dla poszczególnych krajów.
     - `color='continent'`: Kolumna, według której punkty na mapie są kolorowane (np. kontynent).
     - `hover_name='country'`: Nazwa kolumny wyświetlana w interaktywnym podglądzie punktów na mapie (np. nazwa kraju).
     - `size='pop'`: Kolumna zawierająca dane dotyczące wielkości punktów na mapie (np. populacja).
     - `animation_frame='year'`: Kolumna zawierająca informacje o ramce czasowej dla animacji (np. rok).
     - `projection='natural earth'`: Określa projekcję mapy, w tym przypadku 'natural earth'.
2. Wyświetlenie wykresu:
   - Wywołanie tej funkcji automatycznie wyświetli interaktywny wykres z animacją.
"""

px.scatter_geo(df, locations='iso_alpha', color='continent', hover_name='country',
              size='pop', animation_frame='year', projection='natural earth')
```
