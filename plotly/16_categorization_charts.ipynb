```python
"""
Notatka: Załadowanie zestawu danych "titanic" za pomocą biblioteki Seaborn i wyświetlenie początkowych wierszy ramki danych

Ten skrypt zawiera kod do załadowania zestawu danych "titanic" za pomocą funkcji `load_dataset` z biblioteki Seaborn i wyświetlenia początkowych wierszy ramki danych.

1. Załadowanie zestawu danych "titanic":
   - `from seaborn import load_dataset`: Importuje funkcję `load_dataset` z modułu Seaborn.
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `df = load_dataset('titanic')`: Ładuje zestaw danych "titanic" za pomocą funkcji `load_dataset` z modułu Seaborn i przypisuje go do zmiennej `df`.
2. Wyświetlenie początkowych wierszy ramki danych:
   - `df.head()`: Wyświetla pierwsze pięć wierszy ramki danych `df`.
"""

import plotly.express as px
from seaborn import load_dataset

df = load_dataset('titanic')
df.head()
```

```python
"""
Notatka: Tworzenie wykresu punktowego (scatter plot) dla danych dotyczących wieku i opłaty za bilet, z podziałem na płeć i kolorowanie według przeżycia

Ten skrypt zawiera kod do tworzenia wykresu punktowego (scatter plot) dla danych dotyczących wieku i opłaty za bilet, z podziałem na płeć i kolorowanie według przeżycia.

1. Tworzenie wykresu punktowego:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.scatter(df, x='age', y='fare', color='survived', facet_col='sex')`: Tworzy wykres punktowy na podstawie danych z ramki danych `df`. Wiek pasażera reprezentowany jest na osi X, opłata za bilet na osi Y, a punkty są kolorowane według informacji o przeżyciu. Dodatkowo, wykres jest podzielony na kolumny według płci pasażera.
     - Parametry
"""

px.scatter(df, x='age', y='fare', color='survived', facet_col='sex')
```

```python
"""
Notatka: Tworzenie wykresu punktowego (scatter plot) dla danych dotyczących wieku i opłaty za bilet, z podziałem na płeć w wierszach i kolorowanie według przeżycia

Ten skrypt zawiera kod do tworzenia wykresu punktowego (scatter plot) dla danych dotyczących wieku i opłaty za bilet, z podziałem na płeć w wierszach i kolorowanie według przeżycia.

1. Tworzenie wykresu punktowego:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.scatter(df, x='age', y='fare', color='survived', facet_row='sex')`: Tworzy wykres punktowy na podstawie danych z ramki danych `df`. Wiek pasażera reprezentowany jest na osi X, opłata za bilet na osi Y, a punkty są kolorowane według informacji o przeżyciu. Dodatkowo, wykres jest podzielony na wiersze według płci pasażera.
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `x='age'`: Nazwa kolumny reprezentującej wartości na osi X (wiek).
       - `y='fare'`: Nazwa kolumny reprezentującej wartości na osi Y (opłata za bilet).
       - `color='survived'`: Kolumna, według której kolorowane są punkty na wykresie (informacja o przeżyciu).
       - `facet_row='sex'`: Podział wykresu na wiersze według płci pasażera.
"""

px.scatter(df, x='age', y='fare', color='survived', facet_row='sex')
```

```python
"""
Notatka: Tworzenie wykresu punktowego (scatter plot) dla danych dotyczących wieku i opłaty za bilet, z podziałem na płeć i klasę pasażera, kolorowanie według przeżycia

Ten skrypt zawiera kod do tworzenia wykresu punktowego (scatter plot) dla danych dotyczących wieku i opłaty za bilet, z podziałem na płeć i klasę pasażera, kolorowanie według przeżycia.

1. Tworzenie wykresu punktowego:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.scatter(df, x='age', y='fare', color='survived', facet_col='sex', facet_row='pclass', category_orders={'pclass': [3,2,1]})`: Tworzy wykres punktowy na podstawie danych z ramki danych `df`. Wiek pasażera reprezentowany jest na osi X, opłata za bilet na osi Y, a punkty są kolorowane według informacji o przeżyciu. Dodatkowo, wykres jest podzielony na kolumny według płci pasażera i wiersze według klasy pasażera. Klasa pasażera jest sortowana w kolejności [3, 2, 1].
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `x='age'`: Nazwa kolumny reprezentującej wartości na osi X (wiek).
       - `y='fare'`: Nazwa kolumny reprezentującej wartości na osi Y (opłata za bilet).
       - `color='survived'`: Kolumna, według której kolorowane są punkty na wykresie (informacja o przeżyciu).
       - `facet_col='sex'`: Podział wykresu na kolumny według płci pasażera.
       - `facet_row='pclass'`: Podział wykresu na wiersze według klasy pasażera.
       - `category_orders={'pclass': [3,2,1]}`: Określa kolejność klas pasażera na osi wierszy.
"""

px.scatter(df, x='age', y='fare', color='survived', facet_col='sex', facet_row='pclass', category_orders={'pclass': [3,2,1], 'sex': ['female', 'male']})
```
