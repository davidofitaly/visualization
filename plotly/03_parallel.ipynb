```python
"""
Notatka: Tworzenie wykresu równoległych współrzędnych za pomocą funkcji px.parallel_coordinates()

Ten skrypt zawiera kod do tworzenia wykresu równoległych współrzędnych za pomocą funkcji `px.parallel_coordinates()` z różnymi parametrami.

1. Tworzenie wykresu równoległych współrzędnych:
   - `px.parallel_coordinates(data_frame=data, color='species_id', height=800)`
     - Tworzy wykres równoległych współrzędnych na podstawie danych z ramki danych `data`. Wykres ten pokazuje zależności między wartościami różnych wymiarów (kolumn), gdzie każda linia reprezentuje jedną próbkę danych.
     - Parametry:
       - `data_frame=data`: Ramka danych zawierająca dane do wykresu.
       - `color='species_id'`: Kolorowanie linii na wykresie na podstawie identyfikatora gatunku.
       - `height=800`: Wysokość wykresu w pikselach.

"""

import plotly.express as px

data = px.data.iris()
data.head()

px.parallel_coordinates(data,
                        color='species_id',
                        height=800
                        )
```

```python
"""
Notatka: Tworzenie wykresu kategorii równoległych za pomocą funkcji px.parallel_categories()

Ten skrypt zawiera kod do tworzenia wykresu kategorii równoległych za pomocą funkcji `px.parallel_categories()` z różnymi parametrami.

1. Tworzenie wykresu kategorii równoległych:
   - `px.parallel_categories(data_frame=data, color='size', height=800)`
     - Tworzy wykres kategorii równoległych na podstawie danych z ramki danych `data`. Wykres ten pokazuje zależności między kategoriami, gdzie każda linia reprezentuje jedną próbkę danych.
     - Parametry:
       - `data_frame=data`: Ramka danych zawierająca dane do wykresu.
       - `color='size'`: Kolorowanie linii na wykresie na podstawie rozmiaru.
       - `height=800`: Wysokość wykresu w pikselach.
"""

data= px.data.tips()

px.parallel_categories(data,
                       color='size',
                       height=800,
                       )
```
