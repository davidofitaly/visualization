```python
"""
Notatka: Załadowanie zestawu danych "mpg" za pomocą biblioteki Seaborn i wyświetlenie początkowych wierszy ramki danych

Ten skrypt zawiera kod do załadowania zestawu danych "mpg" za pomocą funkcji `load_dataset` z biblioteki Seaborn i wyświetlenia początkowych wierszy ramki danych.

1. Załadowanie zestawu danych "mpg":
   - `from seaborn import load_dataset`: Importuje funkcję `load_dataset` z modułu Seaborn.
   - `import plotly.graph_objects as go`: Importuje moduł Plotly Graph Objects jako `go`.
   - `df = load_dataset('mpg')`: Ładuje zestaw danych "mpg" za pomocą funkcji `load_dataset` z modułu Seaborn i przypisuje go do zmiennej `df`.
2. Wyświetlenie początkowych wierszy ramki danych:
   - `df.head()`: Wyświetla pierwsze pięć wierszy ramki danych `df`.
"""

from seaborn import load_dataset
import plotly.graph_objects as go

df = load_dataset('mpg')
df.head()
```

```python
"""
Notatka: Tworzenie wykresu dwuwymiarowego histogramu oraz dodanie punktów za pomocą biblioteki Plotly Graph Objects

Ten skrypt zawiera kod do tworzenia wykresu dwuwymiarowego histogramu oraz dodania punktów za pomocą obiektów `Histogram2d` i `Scatter` z biblioteki Plotly Graph Objects.

1. Tworzenie wykresu dwuwymiarowego histogramu:
   - `import plotly.graph_objects as go`: Importuje moduł Plotly Graph Objects jako `go`.
   - `fig = go.Figure(go.Histogram2d(`: Tworzy obiekt klasy `Figure`, na którym zostanie umieszczony wykres dwuwymiarowego histogramu.
      - `x = df.mpg`: Dane osi X pochodzą z kolumny 'mpg' ramki danych `df`.
      - `y = df.acceleration`: Dane osi Y pochodzą z kolumny 'acceleration' ramki danych `df`.
      - `nbinsx=30, nbinsy=30`: Liczba przedziałów (binów) dla osi X i Y wynosi 30.
2. Dodanie punktów do wykresu:
   - `fig.add_trace(go.Scatter(`: Dodaje punkty do istniejącego wykresu dwuwymiarowego histogramu.
      - `x = df.mpg`: Współrzędne X punktów pochodzą z kolumny 'mpg' ramki danych `df`.
      - `y = df.acceleration`: Współrzędne Y punktów pochodzą z kolumny 'acceleration' ramki danych `df`.
      - `mode='markers'`: Określa tryb wykresu jako punkty.
      - `marker={...}`: Określa parametry punktów, takie jak symbol, przezroczystość, kolor, rozmiar i grubość linii.
3. Wyświetlenie wykresu:
   - `fig.show()`: Wyświetla wykres.
"""

fig = go.Figure(go.Histogram2d(
    x = df.mpg,
    y = df.acceleration,
    nbinsx=30,
    nbinsy=30
))
fig.add_trace(go.Scatter(
    x = df.mpg,
    y = df.acceleration,
    mode='markers',
    marker={
        'symbol': 'x',
        'opacity': 0.7,
        'color': 'white',
        'size': 8,
        'line': {
            'width': 1
        }
    }
))
fig.show()
```
