```python
"""
Notatka: Wczytywanie danych dotyczących Gapminder

Ten skrypt zawiera kod do wczytywania danych dotyczących Gapminder za pomocą modułu `plotly.express`.

1. Wczytywanie danych:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `df = px.data.gapminder()`: Wczytuje dane dotyczące Gapminder za pomocą wbudowanej funkcji `gapminder()` z modułu Plotly Express i zapisuje je do zmiennej `df`.
   - `df.head()`: Wyświetla pierwsze pięć wierszy wczytanych danych w celu weryfikacji poprawności wczytania.
"""

import plotly.express as px
import pandas as pd
df = px.data.gapminder()
df.head()
```

```python
"""
Notatka: Tworzenie interaktywnej mapy choropleth z danymi na temat oczekiwanej długości życia

Ten skrypt zawiera kod do tworzenia interaktywnej mapy choropleth z danymi na temat oczekiwanej długości życia za pomocą modułu `plotly.express`.

1. Tworzenie interaktywnej mapy choropleth:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.choropleth(`: Tworzy interaktywną mapę choropleth za pomocą funkcji `choropleth` z modułu Plotly Express.
     - `df`: Ramka danych zawierająca dane geograficzne.
     - `locations='iso_alpha'`: Kolumna zawierająca kody ISO dla poszczególnych krajów.
     - `color='lifeExp'`: Kolumna zawierająca dane dotyczące oczekiwanej długości życia, które są używane do kolorowania obszarów na mapie.
     - `hover_name='country'`: Nazwa kolumny wyświetlana w interaktywnym podglądzie obszarów na mapie (np. nazwa kraju).
2. Wyświetlenie mapy:
   - Wywołanie tej funkcji automatycznie wyświetli interaktywną mapę choropleth.
"""

px.choropleth(df, locations='iso_alpha', color='lifeExp', hover_name='country')
```

```python
raw = """AL / Albania
AD / Andorra
AM / Armenia
AT / Austria"
BY / Belarus
BE / Belgium
BA / Bosnia and Herzegovina
BG / Bulgaria
CH / Switzerland
CY / Cyprus
CZ / Czech Republic
DE / Germany
DK / Denmark
EE / Estonia
ES / Spain
FO / Faeroe Islands
FI / Finland
FR / France
GB / United Kingdom
GE / Georgia
GI / Gibraltar
GR / Greece
HU / Hungary
HR / Croatia
IE / Ireland
IS / Iceland
IT / Italy
LT / Lithuania
LU / Luxembourg
LV / Latvia
MC / Monaco
MK / Macedonia
MT / Malta
NO / Norway
NL / Netherlands
PO / Poland
PT / Portugal
RO / Romania
RU / Russian Federation
SE / Sweden
SI / Slovenia
SK / Slovakia (Slovakian Republic)
SM / San Marino
TR / Turkey
UA / Ukraine
VA / Vatican City State
"""
```

```python
"""
Notatka: Przetwarzanie surowych danych z podziałem na kod i nazwę kraju

Ten skrypt zawiera kod do przetwarzania surowych danych z podziałem na kod i nazwę kraju oraz tworzenia ramki danych za pomocą modułu `pandas`.

1. Przetwarzanie surowych danych:
   - `l = raw.split('\n')`: Dzieli surowe dane na wiersze, tworząc listę `l`.
   - `l = [row.strip().split('/') for row in l][:-1]`: Usuwa puste wiersze, dzieli każdy wiersz na elementy za pomocą znaku '/' i usuwa białe znaki z każdego elementu.
2. Tworzenie list kodów i nazw krajów:
   - `code = []` i `country = []`: Inicjują puste listy, które będą przechowywać kody i nazwy krajów.
   - `for item in l:`: Przechodzi przez każdy element listy `l`.
     - `code.append(item[0].strip())` i `country.append(item[1].strip())`: Dodaje odpowiednie kody i nazwy krajów do list `code` i `country`, usuwając ewentualne białe znaki z każdego elementu.
3. Tworzenie ramki danych:
   - `df = pd.DataFrame(dict(code=code, country=country))`: Tworzy ramkę danych `df` z kolumnami `code` i `country` na podstawie list `code` i `country`.
   - `df = df.reset_index()`: Resetuje indeksy ramki danych.
4. Wyświetlenie początkowych wierszy ramki danych:
   - `df.head()`: Wyświetla pierwsze pięć wierszy ramki danych w celu weryfikacji poprawności przetwarzania danych.
"""

l = raw.split('\n')
l = [row.strip().split('/') for row in l][:-1]

code = []
country = []
for item in l:
    code.append(item[0].strip())
    country.append(item[1].strip())

print(code)
print(country)

df = pd.DataFrame(dict(code=code, country=country))
df = df.reset_index()
df.head()
```

```python
"""
Notatka: Tworzenie interaktywnej mapy choropleth na podstawie ramki danych

Ten skrypt zawiera kod do tworzenia interaktywnej mapy choropleth na podstawie danych zawartych w ramce danych za pomocą modułu `plotly.express`.

1. Tworzenie interaktywnej mapy choropleth:
   - `import plotly.express as px`: Importuje moduł Plotly Express jako `px`.
   - `px.choropleth(`: Tworzy interaktywną mapę choropleth za pomocą funkcji `choropleth` z modułu Plotly Express.
     - `df`: Ramka danych zawierająca dane geograficzne.
     - `locations='country'`: Kolumna zawierająca nazwy krajów lub kody ISO dla poszczególnych obszarów.
     - `color='index'`: Kolumna zawierająca dane, które są używane do kolorowania obszarów na mapie (np. wskaźnik, wartość, itp.).
     - `locationmode='country names'`: Określa, że dane w kolumnie 'locations' są nazwami krajów.
     - `hover_name='country'`: Nazwa kolumny wyświetlana w interaktywnym podglądzie obszarów na mapie (np. nazwa kraju).
2. Wyświetlenie mapy:
   - Wywołanie tej funkcji automatycznie wyświetli interaktywną mapę choropleth.
"""

px.choropleth(df, locations='country', color='index', locationmode='country names', hover_name='country')
```
