Notatka 1 Stworzenie wykresu jointplot.

```python
"""
Notatka: Tworzenie wykresu punktowego na podstawie losowych danych

Poniżej znajduje się kod do stworzenia wykresu punktowego na podstawie losowych danych przy użyciu biblioteki Seaborn.

1. Tworzenie wykresu punktowego:
   - `import seaborn as sns`: Importuje bibliotekę Seaborn.
   - `import matplotlib.pyplot as plt`: Importuje moduł pyplot z biblioteki Matplotlib.
   - `import numpy as np`: Importuje bibliotekę NumPy.
   - `import pandas as pd`: Importuje bibliotekę Pandas.
   - `sns.set(style='darkgrid')`: Ustawia styl siatki na ciemny (darkgrid) za pomocą funkcji `set` z Seaborn.
   - `df = np.random.multivariate_normal(mean=[0,1], cov=[(1,-0.4), (-0.4, 1)], size=100)`: Generuje losowe dane wielowymiarowe z rozkładu normalnego o średniej [0,1] i macierzy kowariancji [(1,-0.4), (-0.4, 1)] za pomocą funkcji `multivariate_normal` z NumPy.
   - `df = pd.DataFrame(df, columns=['x', 'y'])`: Tworzy obiekt DataFrame z losowych danych, nadając kolumnom nazwy 'x' i 'y' za pomocą funkcji `DataFrame` z Pandas.
   - `df.head(3)`: Wyświetla pierwsze trzy wiersze utworzonego DataFrame'a.

Kod ten generuje dane losowe oraz tworzy DataFrame na ich podstawie, które są wykorzystywane do stworzenia wykresu punktowego.
"""

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sns.set(style='darkgrid')

df = np.random.multivariate_normal(mean=[0,1], cov=[(1,-0.4), (-0.4, 1)], size=100)
df = pd.DataFrame(df, columns=['x', 'y'])
df.head(3)
```

```python
"""
Notatka: Tworzenie wykresu łącznego (joint plot) na podstawie losowych danych

Poniżej znajduje się kod do stworzenia wykresu łącznego (joint plot) na podstawie losowych danych przy użyciu biblioteki Seaborn.

1. Tworzenie wykresu łącznego:
   - `sns.jointplot(data=df, x='x', y='y')`: Tworzy wykres łączny (joint plot), który łączy wykres punktowy z histogramami na podstawie danych z DataFrame'a `df`. Oś x wykresu punktowego odpowiada kolumnie 'x', a oś y wykresu punktowego odpowiada kolumnie 'y'.

Kod ten tworzy wykres łączny, który wyświetla zarówno rozkład zmiennych 'x' i 'y' w postaci histogramów, jak i ich relację na wykresie punktowym.
"""

sns.jointplot(data=df, x='x', y='y')
```

```python
"""
Notatka: Tworzenie wykresu łącznego (joint plot) z heksagonalnymi bryłami (hexbin) na podstawie losowych danych

Poniżej znajduje się kod do stworzenia wykresu łącznego (joint plot) z heksagonalnymi bryłami (hexbin) na podstawie losowych danych przy użyciu biblioteki Seaborn.

1. Tworzenie wykresu łącznego z heksagonalnymi bryłami:
   - `sns.jointplot(data=df, x='x', y='y', kind='hex')`: Tworzy wykres łączny (joint plot) z heksagonalnymi bryłami (hexbin), który pokazuje rozkład dwóch zmiennych 'x' i 'y' za pomocą heksagonalnych brył na podstawie danych z DataFrame'a `df`.

Kod ten tworzy wykres łączny z heksagonalnymi bryłami, które reprezentują gęstość punktów w obszarach o różnych wartościach 'x' i 'y'.
"""


sns.set(style='white')
sns.jointplot(data=df, x='x', y='y', kind='hex')
```

```python
"""
Notatka: Tworzenie wykresu łącznego (joint plot) z oszacowaniami gęstości jądra (kernel density estimation) na podstawie losowych danych

Poniżej znajduje się kod do stworzenia wykresu łącznego (joint plot) z oszacowaniami gęstości jądra (kernel density estimation) na podstawie losowych danych przy użyciu biblioteki Seaborn.

1. Tworzenie wykresu łącznego z oszacowaniami gęstości jądra:
   - `g = sns.jointplot(data=df, x='x', y='y', kind='kde', nlevels=15, cmap='viridis')`: Tworzy wykres łączny (joint plot) z oszacowaniami gęstości jądra (KDE), który pokazuje rozkład dwóch zmiennych 'x' i 'y' na podstawie danych z DataFrame'a `df`. `nlevels` określa liczbę poziomów konturów, a `cmap` ustawia mapę kolorów.
   - `g.plot_joint(plt.scatter, c='r', s=30, marker='+')`: Dodaje punkty danych na wykresie łącznym w miejscach, gdzie wartości 'x' i 'y' spełniają warunek, który jest zdefiniowany przez funkcję `plt.scatter`. Argumenty `c`, `s` i `marker` określają kolor, rozmiar i styl punktów, odpowiednio.

Kod ten tworzy wykres łączny z oszacowaniami gęstości jądra (KDE) oraz dodaje czerwone krzyże na wykresie łącznym.
"""

g = sns.jointplot(data=df, x='x', y='y', kind='kde', nlevels=15, cmap='viridis')
g.plot_joint(plt.scatter, c='r', s=30, marker='+')
```
