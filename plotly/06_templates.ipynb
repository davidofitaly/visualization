```python
"""
Notatka: Tworzenie wykresu punktowego na podstawie danych z zestawu danych Gapminder

Ten skrypt zawiera kod do tworzenia wykresu punktowego na podstawie danych z zestawu danych Gapminder za pomocą biblioteki Plotly Express.

1. Importowanie biblioteki Plotly Express:
   - `import plotly.express as px`
     - Importuje bibliotekę Plotly Express i nadaje jej alias `px`.

2. Przygotowanie danych:
   - `df = px.data.gapminder()`
     - Pobiera dane o średniej długości życia, PKB per capita i populacji z zestawu danych `gapminder` dostępnego w bibliotece Plotly Express.

3. Wyświetlanie początkowych wierszy danych:
   - `df.head()`
     - Wyświetla pierwsze kilka wierszy danych w celu ich wizualizacji.
"""

import plotly.express as px

df = px.data.gapminder()
df.head()

```

```python
"""
Notatka: Tworzenie wykresu punktowego z różnymi szablonami wyglądu

Ten skrypt zawiera kod do tworzenia wykresu punktowego za pomocą funkcji `px.scatter()` z różnymi szablonami wyglądu dostępnymi w bibliotece Plotly Express.

1. Importowanie biblioteki Plotly Express:
   - `import plotly.express as px`
     - Importuje bibliotekę Plotly Express i nadaje jej alias `px`.

2. Pętla for do tworzenia wykresów dla różnych szablonów:
   - `for template in ['seaborn', 'none', 'plotly_white', 'ggplot2', 'plotly', 'plotly_dark']:`
     - Iteruje przez listę różnych szablonów wyglądu.

3. Tworzenie wykresu punktowego dla każdego szablonu:
   - `fig = px.scatter(df.query('year==2007'), x='gdpPercap', y='lifeExp', size='pop', color='continent', log_x=True, size_max=60, height=700, width=1000, title=f'Template {template}', template=template)`
     - Tworzy wykres punktowy na podstawie danych z ramki danych `df`, gdzie wartości na osi X są pobierane z kolumny `gdpPercap`, a wartości na osi Y są pobierane z kolumny `lifeExp`. Wielkość punktów jest proporcjonalna do wartości z kolumny `pop`, a kolory punktów są określane przez kolumnę `continent`. Wykres ma skalę logarytmiczną na osi X i ustala maksymalną wielkość punktów na 60. Szerokość i wysokość wykresu są ustawione na 1000 i 700 pikseli odpowiednio. Tytuł wykresu zawiera nazwę aktualnie używanego szablonu.
     - Parametry:
       - `df.query('year==2007')`: Wybiera dane tylko dla roku 2007.
       - `x='gdpPercap'`: Nazwa kolumny reprezentującej wartości na osi X (PKB per capita).
       - `y='lifeExp'`: Nazwa kolumny reprezentującej wartości na osi Y (średnia długość życia).
       - `size='pop'`: Nazwa kolumny reprezentującej wielkość punktów na wykresie (populacja).
       - `color='continent'`: Nazwa kolumny reprezentującej kolorowanie punktów na wykresie (kontynent).
       - `log_x=True`: Ustawia skalę logarytmiczną na osi X.
       - `size_max=60`: Ustala maksymalną wielkość punktów na 60.
       - `height=700`: Wysokość wykresu w pikselach.
       - `width=1000`: Szerokość wykresu w pikselach.
       - `title=f'Template {template}'`: Tytuł wykresu zawiera nazwę aktualnie używanego szablonu.
       - `template=template`: Ustawia aktualnie używany szablon wyglądu.
   - `fig.show()`: Wyświetla aktualny wykres.
"""

for template in ['seaborn', 'none', 'plotly_white', 'ggplot2', 'plotly', 'plotly_dark']:
    fig = px.scatter(df.query('year==2007'),
            x='gdpPercap',
            y='lifeExp',
            size='pop',
            color='continent',
            log_x=True,
            size_max=60,
            height=700,
            width=1000,
            title=f'Template {template}',
            template=template # 'plotly_dark', 'plotly', 'ggplot2', 'seaborn','none', 'plotly_white'
            )
    print()
    fig.show()
```

```python
"""
Notatka: Tworzenie wykresu punktowego z domyślnym szablonem wyglądu

Ten skrypt zawiera kod do tworzenia wykresu punktowego za pomocą biblioteki Plotly Express, gdzie domyślny szablon wyglądu jest ustawiany na 'plotly_dark' za pomocą modułu `plotly.io`.

1. Importowanie biblioteki Plotly IO:
   - `import plotly.io as pio`
     - Importuje moduł Plotly IO i nadaje mu alias `pio`.

2. Ustawianie domyślnego szablonu wyglądu:
   - `pio.templates.default = 'plotly_dark'`
     - Ustawia domyślny szablon wyglądu na 'plotly_dark', który będzie stosowany do wszystkich przyszłych wykresów.

3. Przygotowanie danych:
   - `df = px.data.iris()`
     - Pobiera dane dotyczące irysów z zestawu danych `iris` dostępnego w bibliotece Plotly Express.

4. Tworzenie wykresu punktowego:
   - `fig = px.scatter(df, x='sepal_length', y='petal_width')`
     - Tworzy wykres punktowy na podstawie danych z ramki danych `df`, gdzie wartości na osi X są pobierane z kolumny `sepal_length`, a wartości na osi Y są pobierane z kolumny `petal_width`.
     - Parametry:
       - `df`: Ramka danych zawierająca dane do wykresu.
       - `x='sepal_length'`: Nazwa kolumny reprezentującej wartości na osi X (długość działki kielicha).
       - `y='petal_width'`: Nazwa kolumny reprezentującej wartości na osi Y (szerokość płatka).
       - Pozostałe parametry, takie jak `size`, `color`, itp., nie zostały określone, więc przyjmują wartości domyślne.
   - `fig.show()`: Wyświetla utworzony wykres punktowy.
"""

import plotly.io as pio

pio.templates.default = 'plotly_dark'
df = px.data.iris()
fig = px.scatter(df, x='sepal_length', y='petal_width')
fig.show()
```
