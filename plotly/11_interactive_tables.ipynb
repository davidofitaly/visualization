```python
"""
Notatka: Tworzenie tabeli za pomocą biblioteki Plotly Graph Objects

Ten skrypt zawiera kod do tworzenia tabeli za pomocą biblioteki Plotly Graph Objects oraz importuje bibliotekę Seaborn.

1. Tworzenie tabeli:
   - `import plotly.graph_objects as go`: Importuje moduł Plotly Graph Objects jako `go`.
   - `import seaborn as sns`: Importuje moduł Seaborn jako `sns`.
   - `fig = go.Figure(data=[go.Table(header=dict(values=['USER ID', 'Amount'],
                                                 fill_color='lightblue',
                                                 line_color='gray'),
                                     cells=dict(values=[['010', '019', '013', '031'], [100,150,80,120]],
                                                fill_color='lightgreen',
                                                line_color='gray'))],
                    layout=go.Layout(width=500, height=300))`: Tworzy tabelę za pomocą obiektów Plotly Graph Objects.
     - Parametry:
       - `header=dict(values=['USER ID', 'Amount'], fill_color='lightblue', line_color='gray')`: Definiuje nagłówek tabeli z wartościami 'USER ID' i 'Amount', z podanymi kolorami wypełnienia i linii.
       - `cells=dict(values=[['010', '019', '013', '031'], [100,150,80,120]], fill_color='lightgreen', line_color='gray')`: Definiuje komórki tabeli z wartościami i kolorami wypełnienia oraz linii.
       - `layout=go.Layout(width=500, height=300)`: Określa układ tabeli, w tym jej szerokość i wysokość.
   - `fig.show()`: Wyświetla tabelę.

"""

import plotly.graph_objects as go
import seaborn as sns

fig = go.Figure(data=[go.Table(header=dict(values=['USER ID', 'Amount'],
                                           fill_color='lightblue',
                                           line_color='gray'),
                               cells=dict(values=[['010', '019', '013', '031'], [100,150,80,120]],
                                          fill_color='lightgreen',
                                          line_color='gray'))],
                layout=go.Layout(width=500, height=300))

fig.show()
```

```python
df = sns.load_dataset('flights')
df.head(3)
```

```python
df = df[:20]
df.describe()
```

```python
"""
Notatka: Tworzenie tabeli za pomocą biblioteki Plotly Graph Objects

Ten skrypt zawiera kod do tworzenia tabeli za pomocą biblioteki Plotly Graph Objects.

1. Tworzenie tabeli:
   - `import plotly.graph_objects as go`: Importuje moduł Plotly Graph Objects jako `go`.
   - `data = [go.Table(header=dict(values=df.columns,
                             fill_color='lightgreen',
                             line_color='gray'),
                 cells=dict(values=[df.year, df.month, df.passengers],
                            fill_color='lightblue',
                            line_color='gray'))]`: Tworzy dane tabeli, zawierające nagłówek i komórki.
     - Parametry:
       - `header=dict(values=df.columns, fill_color='lightgreen', line_color='gray')`: Definiuje nagłówek tabeli z wartościami kolumn ramki danych `df`, z podanymi kolorami wypełnienia i linii.
       - `cells=dict(values=[df.year, df.month, df.passengers], fill_color='lightblue', line_color='gray')`: Definiuje komórki tabeli z wartościami z kolumn 'year', 'month' i 'passengers' ramki danych `df`, z podanymi kolorami wypełnienia oraz linii.
   - `fig = go.Figure(data=data)`: Tworzy obiekt `Figure` zawierający dane tabeli.
   - `fig.show()`: Wyświetla tabelę.
"""

data = [go.Table(header=dict(values=df.columns,
                             fill_color='lightgreen',
                             line_color='gray'),
                 cells=dict(values=[df.year, df.month, df.passengers],
                            fill_color='lightblue',
                            line_color='gray'))]
fig = go.Figure(data=data)
fig.show()
```
