```python
"""
Notatka: Analiza danych z zestawu danych Gapminder za pomocą biblioteki Plotly Express

Ten skrypt zawiera kod do analizy danych z zestawu danych Gapminder za pomocą biblioteki Plotly Express.

1. Importowanie niezbędnych bibliotek:
   - `import plotly.express as px`: Importuje bibliotekę Plotly Express i nadaje jej alias `px`.

2. Pobieranie danych:
   - `df = px.data.gapminder()`: Pobiera zestaw danych Gapminder i zapisuje go w zmiennej `df`.

3. Wyświetlanie początkowych wierszy danych:
   - `df.head()`: Wyświetla początkowe wiersze zestawu danych Gapminder w celu zapoznania się z jego strukturą i zawartością.
"""

import plotly.express as px

df = px.data.gapminder()
df.head()

```

```python
"""
Notatka: Tworzenie wykresu liniowego przedstawiającego zmiany populacji w Polsce w czasie

Ten skrypt zawiera kod do tworzenia wykresu liniowego przedstawiającego zmiany populacji w Polsce w czasie za pomocą biblioteki Plotly Express.

1. Tworzenie wykresu liniowego:
   - `px.line()`: Tworzy wykres liniowy na podstawie podanych danych.
     - Parametry:
       - `df.query("country == 'Poland'")`: Wybiera tylko dane dotyczące Polski.
       - `x='year'`: Określa, że wartości na osi X będą pobierane z kolumny 'year' (rok).
       - `y='pop'`: Określa, że wartości na osi Y będą pobierane z kolumny 'pop' (populacja).
"""

px.line(df.query("country == 'Poland'"), x='year', y='pop')
```

```python
"""
Notatka: Tworzenie wykresu liniowego przedstawiającego zmiany populacji w krajach europejskich w czasie

Ten skrypt zawiera kod do tworzenia wykresu liniowego przedstawiającego zmiany populacji w krajach europejskich w czasie za pomocą biblioteki Plotly Express.

1. Tworzenie wykresu liniowego:
   - `px.line()`: Tworzy wykres liniowy na podstawie podanych danych.
     - Parametry:
       - `df.query("continent == 'Europe'")`: Wybiera tylko dane dotyczące krajów europejskich.
       - `x='year'`: Określa, że wartości na osi X będą pobierane z kolumny 'year' (rok).
       - `y='pop'`: Określa, że wartości na osi Y będą pobierane z kolumny 'pop' (populacja).
       - `color='country'`: Określa, że każdy kraj będzie miał inny kolor linii na wykresie.
       - `height=800`: Określa wysokość wykresu w pikselach.
       - `width=800`: Określa szerokość wykresu w pikselach.
"""

px.line(df.query("continent == 'Europe'"),
        x='year',
        y='pop',
        color='country',
        height=800,
        width=800)
```
