Notatka 1 Przygotowanie danych.

```python
"""
Notatka: Wczytywanie danych i wyświetlanie podglądu

Poniżej znajduje się kod do wczytania zestawu danych 'tips' za pomocą biblioteki Seaborn oraz wyświetlenia trzech pierwszych wierszy tego zestawu danych.

1. Wczytywanie danych:
   - `import seaborn as sns`: Importuje bibliotekę Seaborn.
   - `import matplotlib.pyplot as plt`: Importuje moduł pyplot z biblioteki Matplotlib, który umożliwia tworzenie wykresów.
   - `df = sns.load_dataset('tips')`: Wczytuje zestaw danych 'tips' z dostępnych zestawów danych w Seaborn i przypisuje go do zmiennej df.

2. Podgląd danych:
   - `df.head(3)`: Wyświetla pierwsze trzy wiersze wczytanego zestawu danych 'tips'.
"""

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')
df.head(3)
```

Notatka 1.1 Stworzenie wykresu sns.catplot().

```python
"""
Notatka: Tworzenie wykresu kategorii (catplot) z ustawieniami stylu i palety kolorów

Poniżej przedstawiono kod do utworzenia wykresu kategorii (catplot) za pomocą biblioteki Seaborn, gdzie wykorzystano określone ustawienia stylu i palety kolorów.

1. Tworzenie wykresu:
   - `sns.set(style='ticks', palette='pastel')`: Ustawia styl wykresu na 'ticks' oraz wybiera paletę kolorów 'pastel', która zawiera pastelowe kolory.
   - `sns.catplot(data=df, x='day', y='total_bill', hue='day')`: Tworzy wykres kategorii na podstawie danych z zestawu df, gdzie na osi x znajduje się kategoria 'day', na osi y wartości 'total_bill', a parametr hue='day' określa, że dane będą podzielone na grupy według kategorii 'day' i każda grupa będzie reprezentowana innym kolorem.

Parametry użyte w funkcji catplot:
   - `data=df`: Dane do wyświetlenia na wykresie.
   - `x='day'`: Zmienna do wyświetlenia na osi x (dzień tygodnia).
   - `y='total_bill'`: Zmienna do wyświetlenia na osi y (łączna kwota rachunku).
   - `hue='day'`: Zmienna do podziału danych na grupy (dzień tygodnia), gdzie każda grupa będzie reprezentowana innym kolorem.

Ustawienia stylu i palety kolorów:
   - `style='ticks'`: Ustawia styl wykresu na 'ticks', co oznacza, że będą widoczne kratki na osiach.
   - `palette='pastel'`: Wybiera paletę kolorów 'pastel', która zawiera pastelowe kolory.
"""

sns.set(style='ticks', palette='pastel')
sns.catplot(data=df, x='day', y='total_bill', hue='day')
```

```python
"""
Notatka: Tworzenie wykresu kategorii (catplot) z ustawieniami stylu, palety kolorów i typu wykresu

Poniżej przedstawiono kod do utworzenia wykresu kategorii (catplot) za pomocą biblioteki Seaborn, gdzie wykorzystano określone ustawienia stylu, palety kolorów i typu wykresu.

1. Tworzenie wykresu:
   - `sns.set(style='ticks', palette='pastel')`: Ustawia styl wykresu na 'ticks' oraz wybiera paletę kolorów 'pastel', która zawiera pastelowe kolory.
   - `sns.catplot(data=df, x='day', y='total_bill', hue='day', kind='swarm')`: Tworzy wykres kategorii na podstawie danych z zestawu df, gdzie na osi x znajduje się kategoria 'day', na osi y wartości 'total_bill', a parametr hue='day' określa, że dane będą podzielone na grupy według kategorii 'day' i każda grupa będzie reprezentowana innym kolorem. Dodatkowo, parametr kind='swarm' określa, że typem wykresu będzie swarm plot, czyli wykres rojowy.

Parametry użyte w funkcji catplot:
   - `data=df`: Dane do wyświetlenia na wykresie.
   - `x='day'`: Zmienna do wyświetlenia na osi x (dzień tygodnia).
   - `y='total_bill'`: Zmienna do wyświetlenia na osi y (łączna kwota rachunku).
   - `hue='day'`: Zmienna do podziału danych na grupy (dzień tygodnia), gdzie każda grupa będzie reprezentowana innym kolorem.
   - `kind='swarm'`: Określa typ wykresu jako swarm plot, czyli wykres rojowy.

Ustawienia stylu i palety kolorów:
   - `style='ticks'`: Ustawia styl wykresu na 'ticks', co oznacza, że będą widoczne kratki na osiach.
   - `palette='pastel'`: Wybiera paletę kolorów 'pastel', która zawiera pastelowe kolory.
"""

sns.set(style='ticks', palette='pastel')
sns.catplot(data=df, x='day', y='total_bill', hue='day', kind='swarm')
```

```python
"""
Notatka: Tworzenie wykresu kropkowego (swarm plot) z użyciem filtra danych

Poniżej przedstawiono kod do utworzenia wykresu kropkowego (swarm plot) za pomocą biblioteki Seaborn, gdzie zastosowano filtrację danych przed wygenerowaniem wykresu.

1. Tworzenie wykresu:
   - `sns.catplot(data=df.query("day == ['Fri', 'Sat', 'Sun']"), x='day', y='total_bill', hue='sex', kind='swarm')`: Tworzy wykres kropkowy (swarm plot) na podstawie danych z zestawu df, gdzie dane zostały wcześniej przefiltrowane za pomocą zapytania query, które wybiera dni tygodnia 'Fri', 'Sat' i 'Sun'. Na osi x znajduje się kategoria 'day' (dzień tygodnia), na osi y 'total_bill' (łączna kwota rachunku), a parametr kind='swarm' określa, że typem wykresu będzie swarm plot. Dodatkowo, parametr hue='sex' powoduje podział danych na serie według wartości w kolumnie 'sex', co oznacza, że każda seria będzie miała inną kolor.

Parametry użyte w funkcji catplot:
   - `data=df.query("day == ['Fri', 'Sat', 'Sun']")`: Dane przefiltrowane, wybierające tylko dni tygodnia 'Fri', 'Sat' i 'Sun'.
   - `x='day'`: Zmienna do wyświetlenia na osi x (dzień tygodnia).
   - `y='total_bill'`: Zmienna do wyświetlenia na osi y (łączna kwota rachunku).
   - `hue='sex'`: Określa, że dane zostaną podzielone na serie według wartości w kolumnie 'sex', co spowoduje, że każda seria będzie miała inną kolor.
   - `kind='swarm'`: Określa typ wykresu jako swarm plot, czyli wykres kropkowy, gdzie punkty danych są rozmieszczone wzdłuż osi kategorialnej w taki sposób, aby unikały nachodzenia na siebie.
"""

sns.catplot(data=df.query("day == ['Fri', 'Sat', 'Sun']"), x='day', y='total_bill', hue='sex', kind='swarm')
```

```python
"""
Notatka: Tworzenie wykresu kategorialnego

Poniżej przedstawiono kod do stworzenia wykresu kategorialnego na podstawie danych o przeżywalności pasażerów statku Titanic.

1. Tworzenie wykresu kategorialnego:
   - `sns.catplot(data=df, x='sex', y='survived', kind='bar', hue='class')`: Tworzy wykres kategorialny na podstawie danych o przeżywalności pasażerów statku Titanic.
     - `data=df`: Dane, na których ma być bazowany wykres.
     - `x='sex'`: Zmienna na osi x, która reprezentuje płeć pasażerów.
     - `y='survived'`: Zmienna na osi y, która reprezentuje informacje o przeżywalności.
     - `kind='bar'`: Określa typ wykresu jako słupkowy.
     - `hue='class'`: Określa dodatkowy wymiar danych (klasa podróżujących), który jest reprezentowany przez różne kolory na wykresie.
"""

sns.catplot(df, x='sex', y='survived', kind='bar', hue='class')
```
