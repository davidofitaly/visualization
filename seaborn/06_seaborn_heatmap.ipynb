```python
"""
Notatka: Tworzenie heatmapy na podstawie losowych danych

1.Tworzenie losowych danych:

uni = np.random.randn(100, 100): Generuje losowe dane numeryczne o rozkładzie normalnym (Gaussa) o wymiarach 100x100 przy użyciu biblioteki NumPy.
Tworzenie heatmapy:

2. sns.heatmap(uni): Tworzy heatmapę na podstawie wygenerowanych danych. Każda komórka heatmapy reprezentuje wartość danych numerycznych, gdzie intensywność koloru wskazuje na wartość liczbową.
Im jaśniejszy kolor, tym większa wartość.

Zastosowanie:

Heatmapa jest używana do wizualizacji danych w postaci kolorowej mapy, gdzie intensywność kolorów reprezentuje wartości numeryczne w określonych punktach.
Jest przydatna do analizy zależności między dwoma zestawami danych oraz do identyfikacji wzorców lub trendów.
"""
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

uni = np.random.randn(100, 100)
sns.heatmap(uni)
```

```python
"""
Notatka: Wczytanie danych i wstępne wyświetlenie

Poniżej znajduje się kod wczytujący dane z zestawu 'flights' za pomocą biblioteki Seaborn oraz wyświetlający kilka pierwszych wierszy z wczytanych danych.

1. Wczytanie danych:
   - `df = sns.load_dataset('flights')`: Wczytuje zestaw danych o lotach ('flights') za pomocą funkcji `load_dataset` z biblioteki Seaborn.

2. Wstępne wyświetlenie danych:
   - `df.head()`: Wyświetla kilka pierwszych wierszy z wczytanych danych, aby zapoznać się z ich strukturą i zawartością.

Wynikiem działania kodu jest wyświetlenie kilku pierwszych wierszy danych z zestawu 'flights'.
"""

df = sns.load_dataset('flights')
df.head()
```

```python
"""
Notatka: Tworzenie heatmapy na podstawie przekształconych danych

Heatmapa jest używana do wizualizacji danych w postaci kolorowej mapy, gdzie intensywność kolorów reprezentuje wartości numeryczne w określonych punktach. Jest przydatna do analizy zależności między dwoma zestawami danych oraz do identyfikacji wzorców lub trendów.

1. Przekształcenie danych:
   - `df = df.pivot(index='month', columns='year', values='passengers')`: Przekształca pierwotny zbiór danych w formę odpowiednią do wygenerowania heatmapy. Dane zostają przekształcone w postać tabeli przestawnej, gdzie indeksem jest miesiąc, kolumnami są lata, a wartościami są liczba pasażerów.

2. Tworzenie heatmapy:
   - `sns.heatmap(df)`: Tworzy heatmapę na podstawie przekształconych danych. Każda komórka heatmapy reprezentuje wartość danych numerycznych, gdzie intensywność koloru wskazuje na wartość liczbową. Im jaśniejszy kolor, tym większa wartość.

Zastosowanie:
-Korelacja między zmiennymi,
Analiza przestrzeni cech.
"""

df = df.pivot(index='month', columns='year', values='passengers')
sns.heatmap(df)
```

```python
"""
Notatka: Generowanie heatmapy na podstawie danych

Poniżej znajduje się kod tworzący heatmapę na podstawie danych z ramki danych `df` za pomocą funkcji `heatmap` z biblioteki Seaborn.

1. Generowanie heatmapy:
   - `sns.heatmap(df, annot=True, fmt='d', cmap='viridis')`: Tworzy heatmapę na podstawie danych z ramki danych `df`.
     - `annot=True`: Włącza wyświetlanie wartości liczbowych w każdej komórce heatmapy.
     - `fmt='d'`: Określa format wyświetlanych liczb jako liczby całkowite.
     - `cmap='viridis'`: Określa mapę kolorów dla heatmapy, w tym przypadku 'viridis'.

Wynikiem działania kodu jest wygenerowanie heatmapy na podstawie danych z ramki danych `df` z wyświetlonymi wartościami liczbowymi w każdej komórce.


"""

sns.heatmap(df,annot=True, fmt='d', cmap='viridis')
```
